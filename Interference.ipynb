## 1. Interference of light: Introduction

<u>Objective:</u>  To provide a foundational understanding of interference within its historical context, enabling a deeper comprehension of the fundamental nature of this phenomenon.

**[Interference](https://en.wikipedia.org/wiki/Wave_interference)** (from Latin <i>interferens</i> is from <i>inter</i> - "between" and <i> ferre </i> - "to carry") is a phenomenon in which two [coherent waves](https://unifyphysics.com/coherent-and-incoherent-addition-of-waves/) combine, taking into account their [phase difference](https://www.nist.gov/image/phase-differencegif). This combination involves adding their displacements, resulting in a wave with either greater intensity (constructive interference) or reduced amplitude (destructive interference), depending on whether the waves are in phase or out of phase, respectively. 

- Interference effects can be observed with all types of waves, for example, light, radio, acoustic, surface water waves, gravity or matter waves.

- The result of interference (interference pattern) depends on the phase difference of the overlapping coherent waves (Fig.1).

Double-slit interference experiment (Young's interference experiment), first conducted by Thomas Young in 1801, is a classic demonstration of the phenomenon of interference. This experiment shows that light behaves as a wave when passing through two closely spaced slits (S1 and S2) with a width comparable to its wavelength. According to Huygens' principle, each slit acts as a source of secondary waves upon interacting with light wave. Because these new light sources are coherent and in-phase, their superposition forms a characteristic interference pattern of bright and dark fringes (Fig. 1) on a screen.

Let us consider a point $P$ on a screen located at a distance $L$ from two parallel slits illuminated by a coherent light wave with a wavelength $\lambda$. Coherence ensures that the light waves maintain a constant phase relationship, which is essential for producing a clear interference pattern.

Assuming that the light from the two sources, $S_1$ and $S_2$, is initially in phase, the phase relationship of the waves at point $P$ will depend on the optical path difference between the rays arriving from each slit: 

$$
 \Delta l= d \sin(\varphi)
$$
where, 
- $d$  is the distance between the centers of the slits;
- $\varphi$  is the angular position of point $P$ with respect to the central axis.

So, the <b>phase difference</b> between the rays coming from the two slits is given by:

$$
 \Delta\phi = \frac{2\pi}{\lambda} \Delta l = \frac{2\pi d}{\lambda} \sin(\varphi) \tag{1}
$$

If the path difference is equal to an integral number of wavelengths the waves will interfere constructively (Fig.1(b)), leading to a bright spot on the screen.  Mathematically we can write this condition for **maximum intensity** as,

$$ d\sin\varphi = \pm m \lambda \tag{2}$$


Similarly, the condition for **minimum intensity** (destructive interference) at point P on the screen, when the path difference between the rays is a multiple of half wavelengths (causing the waves to arrive out of phase), is given by:

$$ d\sin\varphi = \pm (2m + 1)\frac{\lambda}{2}, \tag{3}$$

where $m = 1,2,3...$ order of diffraction. 



### 1.1. Intensity of light in Double-Slit Intereference

To determine how light intensity varies as a function of position on the screen in a double-slit experiment, one must consider the combined effects of <b>interference</b> from the coherent light waves emerging from the two slits and <b>diffraction</b> due to the finite width of each slit.


1. Interference. 

Based on previous descussions the electric fild in point $P$ can be written,

$$ E_p(\varphi) = E_{S_1} + E_{S_2} = E_0\left(\sin \omega t + \sin(\omega t + \Delta \phi)\right) = \\
= 2 E_0 \cos{\frac{\Delta \phi}{2}} \sin(\omega t + \frac{\Delta \phi}{2}).$$


So, intensity observed at $P$ is given by,

$$I_i(\varphi) = 4 E_0^2 \cos^2{\frac{\Delta \phi}{2}}$$


$$ I_i(\varphi) = I_0\cos^2\left(\frac{\pi d}{\lambda} \sin(\varphi) \right) \tag{4}$$

2. Diffraction.
The phenomenon when waves bend around obstacles or propagating through narrow openings, changing direction and deviating from geometric optics predictions, occurs due to the finite width of each slit.
The resulting diffraction pattern can be described using the sinc function, which accounts for the distribution of light due to the slit width:

$$ I_d(\varphi) = I_0  \text{sinc}^2\left(\frac{\pi a}{\lambda} \sin(\varphi)\right) \tag{5}$$

where $a$ is the width of the slit.

3. The overall light intensity $I(\varphi)$ observed at a point P on the screen is the product of the interference and diffraction contributions:

$$I(\varphi) = I_i(\varphi) I_d(\varphi)$$
$$I(\varphi) = I_0\cos^2\left(\frac{\pi d}{\lambda} \sin(\varphi) \right)\text{sinc}^2\left(\frac{\pi a}{\lambda} \sin(\varphi)\right) \tag{6}$$

Thus, the variation in light intensity on the screen in a double-slit experiment is governed by the interplay between interference and diffraction.
Interference between the waves from each slit produces a pattern of bright and dark fringes, while diffraction modulates the envelope of this pattern, affecting the intensity and visibility of the fringes across the screen.
This relationship will be further explored and demonstrated in the next chapter.

###  1.2. Light Interference Modeling

Based on the Huygens-Fresnel principle, when light passes through a double-slit system, each point along the slits acts as a source of secondary spherical waves. If the slits have a width $a$ along the $x$-axis and their size in the other two directions is negligible (creating a simplified two-dimensional system), the electric field at an observation point \(x, y\) can be expressed as the superposition of the contributions from each point on the slits and reflects both the phase difference due to path length and the amplitude from each source, leading to the observed interference pattern.

$$
E(x,z) = \int_{x'}E_1(x,z)dx + \int_{x'}E_2(x, z)dx
$$

with

$$
E_1(x, z) = E_0 \frac{e^{ik\sqrt{(x - x')^2 + z^2}}}{\sqrt{(x - x')^2 + z^2}}
$$

$$
E_2(x, z) = E_0 \frac{e^{ik\sqrt{(x - x')^2 + z^2}}}{\sqrt{(x - x')^2 + z^2}}
$$

and $x' \in (-D/2+a,-D/2) \cup (D/2,D/2+a)$.

where 

$E_0$ is the amplitude of the waves passing through slits,

$D$ is the edge-to-edge distance between the two slits,

$k$ is the wavenumber (related to the wavelength $\lambda$ by $k = \frac{2\pi}{\lambda}$ ).

The intensity of spherical waves at a point \(x, z\) is given by the square modulus of the wave function:

$$
I(x, z) = |E(x, z)|^2 \tag{7}.
$$



#### *Modeling interference pattern produced by a double-slit setup*
The code visualizes the intensity distribution of light waves passing through a double-slit structure in the $(X,Z)$ plane, as well as the intensity projection onto a screen at distance $L$ from the slits. This approach is based on the idea that the resulting wave amplitude at each grid point is determined by superimposing (summing) contributions from each point on the two slits, thereby creating an interference pattern. 

The using formulas (presented above) are valid for the case where the slits are assumed to be infinitely thin along the $y$-axis.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

nm = 1e-9
um = 1e-6
mm = 1e-3
cm = 1e-2
m = 1

# Function to calculate and display wave intensity
def plot_wave_superposition(L=10, a=0.5, D=1, wavelength=0.5):
    
    Lx = 30 # Modeling area along the x-axis
    Lz = 30 # Modeling area along the z-axis
    
    k = 2 * np.pi / wavelength  # Wave Number


    # Creation of the Coordinate Grid
    x = np.linspace(0, Lx, 500)
    z = np.linspace(0, Lz, 500)
    X, Z = np.meshgrid(x, z, indexing="ij")

    # Positions of the Sources
    slits_x = np.hstack((Lx/2 + D/2 + np.linspace(0, a, 30),
                        Lx/2 - D/2 - np.linspace(0, a, 30)))

    # Initialize the superposition
    wave_superposition = 0

    for slit_x in slits_x:
        # Calculation of the Distances from the Grid Points to the Sources
        R = np.sqrt((X - slit_x) ** 2 + (Z - 0) ** 2)

        # Avoid division by zero by adding a small value to R
        wave = np.exp(1j * (k * R)) / (R + 1e-10)

        # Superposition of sinusoidal monochromatic waves
        wave_superposition += wave

    # Determine a fixed value of x to find the intensity distribution at thit position
    z_screen = L
    index_z_screen = np.argmin(np.abs(z - z_screen))


    # Calculate intensity
    intensity = np.abs(wave_superposition)**2    
    # intensity = np.real(wave_superposition)

    # Normalize intensity for better visualization
    intensity = intensity / np.abs(intensity).max()

    # Calculate the intensity intensity on a screen for a fixed L.
    intensity_z_screen = intensity[:, index_z_screen]

    # Create a figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7), gridspec_kw={'width_ratios': [2, 1]})

    # Visualization of wave superposition in all the slits
    im = ax1.imshow(intensity, extent=[0, Lz, 0, Lx], origin='lower', cmap='magma', vmax=intensity_z_screen.max()
    )
    ax1.set_title("Intensity")
    ax1.set_xlabel('z')
    ax1.set_ylabel('x')
    fig.colorbar(im, ax=ax1)

    # Visualization of the intensity on a screen for a fixed L.
    ax2.plot(x, intensity_z_screen, color='blue')
    ax2.set_title(f'Intensity at L(z) = {z_screen}')
    ax2.set_xlabel('x')
    ax2.set_ylabel('Intensity')
  

# Utilisation de interact pour créer les widgets de contrôle
interact(plot_wave_superposition,
            L=widgets.FloatSlider(min=5, max=30, step=1, value=10, description='L'),
            a=widgets.FloatSlider(min=0, max=4, step=0.1, value=0.5, description='a'),
            D=widgets.FloatSlider(min=0, max=50, step=0.1, value=1, description='D'),

            wavelength=widgets.FloatSlider(min=0.1, max=2, step=0.1, value=0.5, description='Wavelength, mkm'),
        )

```

#### *Modeling interference pattern on the screen produced by a double-slit setup*

In this block, the interference pattern formed by light passing through a double-slit setup is modeled using both analytical equations (4)-(6) and a numerical method via FFT. The calculations produce three visual outputs: the geometry of the slits (top), the interference pattern on the screen (middle), and the light intensities on the screen as a function of distance from the central maximum (bottom), obtained by different methods. These figures are generated using the show_diffraction function (for details, see the Sheet class description).

```python
import numpy as np
from scipy.fftpack import fft2, fftshift
import matplotlib.pyplot as plt

nm = 1e-9
um = 1e-6
mm = 1e-3
cm = 1e-2
m = 1
```

```python
class Sheet():
    def __init__(self, L_x, L_y, Nx, Ny, x0 = 0, y0 = 0):
        """
        L_x - calculation rectangular space width along the x-axis
        L_y - calculation rectangular space width along the y-axis
        x0, y0 - centres of the calculation space
        Nx, Ny - the numbers of points of rectangular space along X and Y axes
        """
        self.Nx = int(Nx)
        self.Ny = int(Ny)

        self.x = x0 + np.linspace(-L_x/2, L_x/2, self.Nx)
        self.y = y0 + np.linspace(-L_y/2, L_y/2, self.Ny)
        self.xx, self.yy = np.meshgrid(self.x, self.y)

        self.mask = np.zeros((self.Ny, self.Nx))


    def rectangular_slit(self, x0, y0, lx, ly):
        """
            Creates a slit centered at the point (x0, y0) with width lx and height ly.

            Parameters:
            - x0, y0: Center coordinates of the slit.
            - lx: Width of the slit in the x-direction.
            - ly: Height of the slit in the y-direction.
        """
        # Apply the slit to the E field using a mask
        mask = ((self.xx > (x0 - lx/2)) & (self.xx < (x0 + lx/2)) &
                (self.yy > (y0 - ly/2)) & (self.yy < (y0 + ly/2)))

        self.mask += np.where(mask, 1, 0)

    def show_diffraction(self, Intensity2D, Intensity0_x, x_screen_display, y_screen_display, x_screen, y_screen, z):
        """
            Plotting a figure consisting of three subplots:
                ax1 -  plot of the shape of the slit;
                ax2 -  plot of the 2D distribution of the intensity of the diffraction pattern in the XY plane;
                ax3 -  plot of the intensity of light on the screen as a function of distance from the central maximum (when φ=0) 
            
            Parameters:
            - Intensity2D: the 2D distribution of the intensity of the diffraction pattern in the XY plane
            - Intensity0_x: the intensity of light on the screen as a function of distance from the central maximum (when φ=0)  
            - x_screen_display: displayed area on the screen along the x-axis 
            - y_screen_display: displayed area on the screen along the y-axis 
            - x_screen: screen x-coorinates
            - y_screen: screen y-coorinates
            - z - distance between slits and screen of oservation
         
        """
        #plot with matplotlib
        fig = plt.figure(figsize=(5, 7))

        ax1 = fig.add_subplot(3,1,1)
        ax3 = fig.add_subplot(3,1,3)
        ax2 = fig.add_subplot(3,1,2,sharex=ax3)

        # display slit  configuration
        ax1.imshow(self.mask, aspect="equal", extent=[min(self.x), max(self.x), min(self.y), max(self.y)])
        ax1.set_xlabel("x (m)")
        ax1.set_ylabel("y (m)")

        #display intensity distribution on the screen
        ax2.imshow(Intensity2D, aspect="equal", extent = [x_screen[0], x_screen[-1], y_screen[0], y_screen[-1]], cmap ='gray')
        ax2.set_ylabel("y (m)")
        ax2.set_ylim([-y_screen_display / 2, y_screen_display / 2])
        ax2.set_xlim([-x_screen_display / 2, x_screen_display / 2])

        # display intensity distribution on the screen I(x, y = 0)
        Intensity_x = Intensity2D[self.Ny//2]
        ax3.plot(x_screen, Intensity_x, linewidth = 1, label="Intensity")
        #calculation using analytical formula (subplot ax3, red line)
        if Intensity0_x is not None:
            ax3.plot(x_screen, Intensity0_x, ":", label="Intensity_0")
            ax3.legend()
        ax3.set_xlabel("x (m)")
        ax3.set_ylabel("Intensity (a.u.)", labelpad=20)
        # ax3.set_yticklabels([])
        ax3.set_xlim([-x_screen_display / 2, x_screen_display / 2])


        # Hide x-axis tick labels for ax2
        plt.setp(ax2.get_xticklabels(), visible=False)

        # Adjust figure to correct the figure
        data_aspect_ratio1 = (max(self.y) - min(self.y)) / (max(self.x) - min(self.x))
        data_aspect_ratio2 = (z/2) / (z)

        pos3 = ax3.get_position()  # Get the original position of ax3

        fig_width, fig_height = fig.get_size_inches()  # Figure dimensions in pixels
        fig.set_size_inches((fig_width, int((data_aspect_ratio1 + data_aspect_ratio2 + 0.25) * fig_width)))

        y3 = pos3.y0
        ax3.set_position([pos3.x0, y3, pos3.width, pos3.width *0.25])
        y3 += 0.01 + pos3.width * 0.25
        ax2.set_position([pos3.x0, y3, pos3.width, pos3.width * data_aspect_ratio2])
        y3 += 0.01 + pos3.width * data_aspect_ratio2 + pos3.y0
        ax1.set_position([pos3.x0, y3, pos3.width, pos3.width * data_aspect_ratio1])

        plt.show()
```

```python
#calculation using analytical formulas
def diffraction_Franhover(k, slit_width, slit_height, x_screen, y_screen, z):
    X, Y =  np.meshgrid(x_screen, y_screen)
    beta_x = (k * slit_width / 2) * X/z + 1e-10
    beta_y = (k * slit_height / 2) * Y/z + 1e-10
    # Calculate the diffraction pattern for the single slit
    return (np.sin(beta_x) / beta_x) ** 2 * (np.sin(beta_y) / beta_y) ** 2

def interference_Franhover(k, D, slit_width, slit_height, x_screen, y_screen, z):
    X, Y =  np.meshgrid(x_screen, y_screen)
    # Calculate the interference pattern for the double slit
    alpha = (k * D / 2) * X / z
    interference_pattern = (np.cos(alpha)) ** 2
    return interference_pattern

def double_slit_Franhover(k, D, slit_width, slit_height, x_screen, y_screen, z):
    X, Y =  np.meshgrid(x_screen, y_screen)
    beta_x = (k * slit_width / 2) * X/z + 1e-10
    beta_y = (k * slit_height / 2) * Y/z + 1e-10
    single_slit_diffraction = (np.sin(beta_x) / beta_x) ** 2 * (np.sin(beta_y) / beta_y) ** 2
    
    # Calculate the interference pattern for the double slit
    alpha = (k * D / 2) * X / z
    interference_pattern = (np.cos(alpha)) ** 2
    
    # Combine the single-slit diffraction pattern with the interference pattern
    return single_slit_diffraction * interference_pattern


#numerical calculation: using Fast Fourier Transform (eq.11)
def diffraction_FFT(k, sheet, z):
    intensity = fftshift(fft2(sheet.mask * np.exp(1j * k/(2*z) * (sheet.xx**2 + sheet.yy**2))))
    intensity = np.absolute(intensity)**2
    return intensity / intensity.max()
```

```python
"""
    Simulation input
"""
# calculation area (sheet) around slit in (m)
Lx = 10*2000 *4 * nm
Ly = 10*600*4 * nm
# number of point on screen (for FFT it is better to use odd sampling)
Nx = 151
Ny = 81

# distance from slit to the screen (m)
z = 0.001 * m
# observation area in screen
screen_width = z
screen_height = z

# distance between the slits, if the calculation for two slits
D = 2000*8 * nm

# wavelength (m)
λ = 700 * nm
k = 2 * np.pi / λ

sheet = Sheet(L_x=Lx, L_y=Ly, Nx=Nx, Ny=Ny)

# Create a rectangular slit at the center
slit_width = 2000*4 * nm
slit_height = 300*4 * nm

sheet.rectangular_slit(x0 = -D/2, y0 = 0, lx = slit_width, ly = slit_height)
sheet.rectangular_slit(x0 = +D/2, y0 = 0, lx = slit_width, ly = slit_height)

#screen size mm
dx_screen = z*λ/Lx
dy_screen = z*λ/Ly
x_screen = np.linspace(-dx_screen * ((Nx if Nx % 2 else Nx + 1) // 2), dx_screen * ((Nx if Nx % 2 else Nx - 1) // 2), Nx)
y_screen = np.linspace(-dy_screen * ((Ny if Ny % 2 else Ny + 1) // 2), dy_screen * ((Ny if Ny % 2 else Ny - 1) // 2), Ny)

#calculation using analytical formula (subplot ax3, red line)
intensity_diffr = diffraction_Franhover(k, slit_width, slit_height, x_screen, y_screen, z)
intensity_interf = interference_Franhover(k, D, slit_width, slit_height, x_screen, y_screen, z)
intensity = double_slit_Franhover(k, D, slit_width, slit_height, x_screen, y_screen, z)

#numerical calculation using Fast Fourier Transform (subplot ax3, black line)
intensity_FFT = diffraction_FFT(k, sheet, z)


# Create a single sheet that acts as a double slit
# Width should accommodate both slits and the distance D between them
slit1 = Sheet(L_x=slit_width, L_y=slit_height, Nx=Nx, Ny=Ny, x0=-D/2)
slit2 = Sheet(L_x=slit_width, L_y=slit_height, Nx=Nx, Ny=Ny, x0=D/2)

# Define the slit and screen coordinates
x_screen = np.linspace(-screen_width / 2, screen_width / 2, Nx)
y_screen = np.linspace(-screen_height / 2, screen_height / 2, Ny)
```

```python
# display intencities
sheet.show_diffraction(intensity_FFT, intensity_diffr[sheet.Ny//2], screen_width, screen_height, x_screen, y_screen, z)
sheet.show_diffraction(intensity, intensity_interf[sheet.Ny//2], screen_width, screen_height, x_screen, y_screen, z)
```

### 1.3. Light Interference Applications

Wave interference has a broad range of applications, extending from scientific research to practical technologies. In all these cases, the key component is the [interferometer](https://www.rp-photonics.com/interferometers.html) - a device that extracts necessary information from interference patterns. Let’s focus on one of the types: the Michelson interferometer.

The setup consists of the following components  Fig.2(a):

- A semi-reflective mirror, known as a beam splitter, which divides the beam from a coherent radiation source (such as a laser) into two separate beams.
- Beam 1 reflects off Mirror 1 (which can be rotated) and then crosses the beam splitter again before reaching the photodetector (PD).
- Beam 2 passes through the beam splitter, reflects off Mirror 2 (which can be moved along the system's axis), and is then reflected by the beam splitter before reaching the photodetector.
- The ability to adjust the mirrors allows for control over both the path difference between the beams and the direction of their propagation.

Even the slightest changes in the system - such as adjustments to the length of an arm or variations in the refractive index -cause phase shifts between the optical paths, resulting in different interference patterns:

- **Constructive** interference occurs when the path difference corresponds to a whole number of wavelengths (0, $\lambda$, $2\lambda$…), producing a strong signal.
- **Destructive** interference occurs when the path difference is an odd multiple of half wavelengths ($\lambda/2$, $3\lambda/2$…), resulting in a near-zero signal at the photodetector.
- If the two mirrors are strictly perpendicular to each other, the interference pattern appears as concentric circles, known as fringes of equal inclination . 
- If the mirrors are not perfectly perpendicular, the pattern changes to fringes of equal thickness, forming parallel stripes.

This configuration allows the interferometer to detect and display even minute changes in the optical path, making it highly sensitive to variations within the system.

#### *Modeling of interference pattern produced by Michelson interferometr*

This section proposes modeling the resulting interference pattern formed in the Michelson interferometer (shown schematically in Fig. 2(a)) as a function of system parameters, such as the positions of the mirrors and the refractive index of the medium through which the beam passes. To accomplish this, it is necessary to know the intensity distribution at the screen (observation points). When the intensities of the two superimposed waves are equal, the resulting intensity is given by:

$$I = 2  I_0  (1 + \cos(\Delta \phi)),$$

where $\Delta \phi = \Delta \phi_{tilt} + \Delta \phi_{shift}$:

$$\Delta \phi_{tilt} = \frac{2 \pi} {\lambda} X \theta, $$

$$\Delta \phi_{shift} = \frac{4 \pi} {\lambda} (L_2 n_2 - L_1 n_1) \frac{1 - r^2}{2 ((2L_1 n_1 + L) (2L_11 n_2 + L))},$$

where
$L_1, L_2, L$ -  the distances between M1–BS, M2–BS and Source–BS, respectively.

$\theta$ - the tilt angle of Mirror 1, measured in radians;

$r = \sqrt{X^2 + Y^2}$ - specifies the position of a point $(X,Y)$ on the screen.

## 2. Matter Interference

The double-slit experiment demonstrates that particles, such as electrons or photons, passing through two slits create an interference pattern on a screen, proving that they behave as waves. This experiment in modern physics is a key demonstration of wave-particle duality - a fundamental concept in quantum mechanics. The hypothesis of the dual nature of waves and matter was first proposed by De Broglie in 1924, suggesting that each particle with momentum $p$ has a corresponding wavelength $\lambda$, related to its momentum by the same equation as for photons:

$$\lambda = \frac{h}{p}\tag{8}$$

where $h$ is Planck's constant.

This type of experiment was first performed by Thomas Young in 1801, as a demonstration of the wave behavior of visible light. In 1927, Davisson and Germer demonstrated that electrons exhibit the same behavior, which was later extended to atoms and molecules, thus confirming de Broglie's hypothesis.




### 2.1. The Electron Interference

Young's double-slit experiment, initially conducted with light, has been revisited and extended to various particles, most famously electrons. In 1961, Claus Jönsson performed the first successful double-slit experiment with electrons. In his setup, electrons were accelerated from a source to an energy of 50 keV. The slits, which were extremely narrow (about 0.5 μm) and spaced approximately 0.1 μm apart, were crafted on a thin copper foil. With this arrangement, Jönsson was able to observe clear interference patterns, demonstrating the wave-like behavior of electrons (see Fig.4).

Let $\Psi$ represent the wave function of a particle. When the particle passes through a Young's double-slit obstacle, its wave function splits and interferes with itself. This self-interference can be constructive or destructive, leading to regions of high and low probability of detecting the particle. The probability density at any point is proportional to $∣\Psi∣^2$, which describes the likelihood of finding the particle at that location.

When the particle encounters a detection screen, it localizes at a specific point according to this probability distribution. By sending many particles through the slits, the characteristic interference pattern gradually emerges on the screen.

One of the experimental limits in quantum mechanics is that any measurement can disturb the system being observed. This principle applies here: any attempt to determine which slit the particle passes through disturbs its quantum state, causing the interference pattern to disappear. This phenomenon, known as the "collapse of the wave function," highlights a key feature of quantum measurement.

### 2.2. Simulation of matter waves

<u>Objective:</u> To develop a deeper understanding of the underlying physics of electron diffraction by simulating electron diffraction patterns and analyzing the effects of various parameters on the temporal evolution of electrons in space.

The temporal evolution of electrons in space is described of time-dependent Schrodinger equation:
$$
i\hbar {\frac {d}{dt}} \Psi ={\hat {H}} \Psi, \tag{9}
$$
The solution of Eq.(9) can be written in operator form:
$$
\Psi(t) = e^{ -\frac{i}{\hbar} {\hat {H}} t} \Psi(t=0), \tag{10}
$$
where $\Psi(t)$ is the time evolution of the wave function, starting from the initial state $\Psi(t=0)$.

The exponential operator in Eq.(10) can be expanded in terms of Chebyshev polynomials. The Chebyshev expansion method is often used to approximate functions of operators, especially for large-scale problems, due to its numerical efficiency and stability. The core idea is to expand the exponential operator as a series of Chebyshev polynomials $T_n(x)$, where $x$ is a rescaled version of the Hamiltonian $\hat{H} / \hbar$, ensuring that the spectrum of $x$ lies in the interval $[-1, 1]$, which is the domain of Chebyshev polynomials.

The function $f(x) = e^{-i x t}$ can be expanded in terms of Chebyshev polynomials $T_n(x)$:
$$
   f(x) = e^{-i x t} \approx \sum_{n=0}^{N} c_n T_n(x),
$$

where $T_n(x)$ are Chebyshev polynomials of the first kind, and the coefficients $c_n$ are given by:
$$
   c_n = \frac{2}{\pi} \int_{-1}^{1} \frac{e^{-i t x} T_n(x)}{\sqrt{1 - x^2}} dx = i^n J_n(t),
$$
where $J_n(t)$ are the Bessel functions of the first kind.

Once the expansion coefficients $c_n$ are determined, the time-evolved state can be computed as a linear combination of Chebyshev polynomials applied to the rescaled Hamiltonian $\hat{H}$:
$$
   \Psi(t) \approx \sum_{n=0}^{N} c_n T_n(\hat{H}) \Psi(0).
$$
   
The action of the Chebyshev polynomials $T_n(\hat{H})$ on the initial state $\Psi(0)$ is computed recursively using the recurrence relation for Chebyshev polynomials:
   
$$
   T_0(x) = 1, \quad T_1(x) = x, \quad T_{n+1}(x) = 2x T_n(x) - T_{n-1}(x).
$$
   
So, starting with $\Psi_0 = \Psi(t=0)$ and $\Psi_1 = \hat{H} \Psi_0$, we can compute $\Psi_n = T_n(\hat{H}) \Psi_0$ recursively for $n = 2, 3, \dots, N$.

In practice, the series is truncated at a finite number $N$, which is determined by the desired accuracy and the properties of the Hamiltonian and time evolution. The error decreases exponentially with the number of terms due to the rapid convergence of the Chebyshev expansion for smooth functions.

The initial state $\Psi(0)$ can be presentred int the form Gaussian wave packet $\Psi(t=0) \propto \exp\left({-(x/x_0)^2 - (y/y_0)^2}\right)$.

```python
def Get_Propagator_Chebyshev_Expansion(Psi,H,a,z):
	"""
		This fucntion advances Psi in time units of delta t by 
		the Chebyshev expansion of the propagator.
	
	"""
	Hz = H*z
	g0,g1 = Psi, np.dot(Hz,Psi)
	Psi = a[0]*g0 + a[1]*g1

	for jt in range(2,len(a)):
		g0,g1 = g1, g0 + np.dot(Hz,g1)*2 # reccurence of Chebyshev polynomials
		Psi += a[jt]*g1 # accumulation step

	Psi1 = np.dot(H,Psi)
	return Psi,Psi1

def Get_Chebyshev_Coefficients(Nc,Nc_max,amin,r):
	a=np.zeros(Nc_max,float)
	C=2
	a[0] = special.jv(0,r)
	k = np.arange(1,Nc)
	a[k] = C*special.jv(k,r)

	while (a[Nc-1] > amin):
		a[Nc] = C*special.jv(Nc,r)
		Nc+=1

		if(Nc>Nc_max):
			print('The argument of the Chebyshev exp. is too large')
			break
	return a[:Nc],Nc


def Get_Initial_Wave_Packet2D(x,y,k0x,a0x,k0y,a0y):
	Ax = (2/(np.pi*a0x**2))**(1/4)*np.exp(-(x/a0x)**2)*np.exp(ci*k0x*x)
	Ay = (2/(np.pi*a0y**2))**(1/4)*np.exp(-(y/a0y)**2)*np.exp(ci*k0y*y)
	Psi=	np.einsum( 'x,y->xy',Ax,Ay )
	return Psi

# FFT for single eigen state ( psi(x,y) )
def Get_Fourrier_Transform_WF2Dsingle(psi,x,delta_x,y,delta_y,kx,ky):
	KX,X = np.meshgrid(kx,x, sparse=False, indexing='ij')
	KY,Y = np.meshgrid(ky,y, sparse=False, indexing='ij')

	psi_FT = ( (np.exp(-ci*(KX*X)) * delta_x) @ psi @  (np.exp(-ci*(KY*Y))*delta_y).T )/(2*np.pi) # Integrate Trapeze
	return psi_FT

def Get_Kinetic_Operator(x,mu,LBOXx,Type):
	""" 
		Calculation of the kinetic operator

		inputs:
			x -	vector of  coordinate
			mu -mass
			LBOXx- the length of simulation box
			Type - type of the discretization for the kinetic operator

		Outputs:
			T - 2D array of the kinetic operator 
	
	"""
	N = len(x)
	if(Type==0):
		coef= (np.pi/LBOXx)**2/mu
		Di	= np.arange(1,N)
		if(N%2==0):
			T_diag		= (N*N+2)/6.
			T_nondiag	= (-1.)**Di/(np.sin(Di*np.pi/N))**2
			#print('<br>even T=',coef*hstack( (T_diag, T_nondiag) ))
		else:
			T_diag		= (N*N-1)/6.
			T_nondiag	= (-1.)**Di*np.cos(Di*np.pi/N)/(np.sin(Di*np.pi/N))**2
		T = toeplitz( coef*np.hstack( (T_diag, T_nondiag) ) )
	elif(Type==1):
		# np.sin basis set
		coef	= (np.pi/LBOXx)**2/(2*mu)/2
		T = np.zeros((N,N),float)
		I1, I2 = np.meshgrid(np.arange(N),np.arange(N))
		cind = np.logical_not(I1==I2)
		DI = I1[cind]-I2[cind]
		SI = I1[cind]+I2[cind]
		T[cind] = coef*(-1.)**DI*(1./np.sin(np.pi*DI/(2*(N+1)))**2 - 1./np.sin(np.pi*(SI+2)/(2*(N+1)))**2)
		cind = (I1==I2)
		T[cind] = coef*((2*(N+1)**2+1)/3. - 1/np.sin(np.pi*(I1[cind]+1)/(N+1))**2)
	return T


def Get_Kinetic_Operator2D(x,y,mu,LBOXx,LBOXy,Type=0):
	return Get_Kinetic_Operator(x,mu,LBOXx,Type), Get_Kinetic_Operator(y,mu,LBOXy,Type)


def Get_Hamiltonian_Operator2D(Nx,Ny,Tx,Ty,V):

	return   (np.einsum('ij,kl->ikjl', Tx, np.eye(Ny,Ny))+\
			np.einsum('ij,kl->ljki', Ty, np.eye(Nx,Nx))).reshape((Nx*Ny,Nx*Ny))+\
			np.diag(V.flatten()) # H[xy,xy]


def Get_E_min_max2D(V,delta_x,delta_y,mu):
	pxmax = np.pi/delta_x.min()
	pymax = np.pi/delta_y.min()
	Emax = (pxmax**2+pymax**2)/(2.*mu)+V.max()
	Emin = V.min()
	Eshift = (Emax+Emin)/2

	return Emax,Emin,pxmax,pymax,Eshift

def Get_Constant_K(Nk,delta_x):
	kmax = np.pi/delta_x.max()
	k = np.linspace(-kmax,kmax+kmax/Nk, num=2*Nk+1, endpoint=False)
	dx = np.diff(k)
	delta_k = (np.concatenate([[0],dx]) + np.concatenate([dx,[0]]))/2										
	return k,delta_k,kmax

def Get_Constant_K2D(Nkx,delta_x,Nky,delta_y):
	kx,delta_kx,kxmax = Get_Constant_K(Nkx,delta_x)
	ky,delta_ky,kymax = Get_Constant_K(Nky,delta_y)

	return kx,delta_kx,kxmax,ky,delta_ky,kymax
```

```python
def Get_Profile(InputData):
	"""
		Calculate potential energy profile. 

		input:
		 Fields of InputData
		
		output:
		 Fields of OutData argument :
			V - Vector of the potential energy data
			x - x-coordinate
			y - y-coordinate
	"""
	OutData				= lambda:0

	InputData.xc = 0#(InputData.xL+InputData.x0)/2
	InputData.yc = (InputData.yL+InputData.y0)/2
	
	# Profile type

	if(InputData.Pot_Type==20):
		InputData.xmin_g,InputData.xmax_g	= InputData.x0,InputData.xL
		InputData.ymin_g,InputData.ymax_g	= InputData.y0,InputData.yL
		
		OutData.x,InputData.delta_x0,InputData.delta_x,OutData.y,InputData.delta_y0,InputData.delta_y,InputData.LBOXx,InputData.LBOXy	= Get_Constant_Grid2D(InputData.Nx,InputData.x0,InputData.xL,InputData.Ny,InputData.y0,InputData.yL,InputData.Type)
			
		#   Build the potential 
		OutData.V = Get_Potential_Operator2D(OutData.x,OutData.y,None,InputData.Pot_Type)

	elif(InputData.Pot_Type==31): # 1 slit	
		InputData.xmin_g,InputData.xmax_g	= InputData.x0,InputData.xL
		InputData.ymin_g,InputData.ymax_g	= InputData.y0,InputData.yL

		OutData.x,InputData.delta_x0,InputData.delta_x,OutData.y,InputData.delta_y0,InputData.delta_y,InputData.LBOXx,InputData.LBOXy	= Get_Constant_Grid2D(InputData.Nx,InputData.x0,InputData.xL,InputData.Ny,InputData.y0,InputData.yL,InputData.Type)
		
		#  Build the potential 
		OutData.V = Get_Potential_Operator2D(OutData.x,OutData.y,{'V0':InputData.V0,'l':InputData.slit_d,'ls':InputData.slit_ls,'xc':InputData.xc,'yc':InputData.yc},InputData.Pot_Type)

	elif(InputData.Pot_Type==32): # 2 slits
		InputData.xmin_g,InputData.xmax_g	= InputData.x0,InputData.xL
		InputData.ymin_g,InputData.ymax_g	= InputData.y0,InputData.yL

		OutData.x,InputData.delta_x0,InputData.delta_x,OutData.y,InputData.delta_y0,InputData.delta_y,InputData.LBOXx,InputData.LBOXy	= Get_Constant_Grid2D(InputData.Nx,InputData.x0,InputData.xL,InputData.Ny,InputData.y0,InputData.yL,InputData.Type)
		
		#  Build the potential 
		OutData.V = Get_Potential_Operator2D(OutData.x,OutData.y,{'V0':InputData.V0,'l':InputData.slit_d,'ls':InputData.slit_ls,'ly':InputData.slit_ly,'xc':InputData.xc,'yc':InputData.yc},InputData.Pot_Type)

	elif(InputData.Pot_Type==33): # N slits
		if(InputData.n_slits * InputData.slit_ls + (InputData.n_slits - 1+2) * InputData.slit_ly >= (InputData.yL - InputData.y0)):
			sys.exit('Number of slits too large for the current grid!!!') 

		print("Get_Profile",InputData.n_slits)

		InputData.xmin_g,InputData.xmax_g	= InputData.x0,InputData.xL
		InputData.ymin_g,InputData.ymax_g	= InputData.y0,InputData.yL

		OutData.x,InputData.delta_x0,InputData.delta_x,OutData.y,InputData.delta_y0,InputData.delta_y,InputData.LBOXx,InputData.LBOXy	= Get_Constant_Grid2D(InputData.Nx,InputData.x0,InputData.xL,InputData.Ny,InputData.y0,InputData.yL,InputData.Type)

		#  Build the potential 
		OutData.V = Get_Potential_Operator2D(OutData.x,OutData.y,{'V0':InputData.V0,'l':InputData.slit_d,'ls':InputData.slit_ls,'ly':InputData.slit_ly,'n_slits':InputData.n_slits,'xc':InputData.xc,'yc':InputData.yc},InputData.Pot_Type)
	
	elif(InputData.Pot_Type==34): # g	
		InputData.xmin_g,InputData.xmax_g	= InputData.x0,InputData.xL
		InputData.ymin_g,InputData.ymax_g	= InputData.y0,InputData.yL

		OutData.x,InputData.delta_x0,InputData.delta_x,OutData.y,InputData.delta_y0,InputData.delta_y,InputData.LBOXx,InputData.LBOXy	= Get_Constant_Grid2D(InputData.Nx,InputData.x0,InputData.xL,InputData.Ny,InputData.y0,InputData.yL,InputData.Type)
		
		#  Build the potential 
		OutData.V = Get_Potential_Operator2D(OutData.x,OutData.y,{'g':InputData.g,'xc':InputData.xc,'yc':InputData.yc},InputData.Pot_Type)

	else:
		OutputData.Error = 'The type of potential energy profile is not correct (Pot_Type = %i)!'%(InputData.Pot_Type) # type: ignore

	return OutData,InputData
```

```python
"""
	Enter your data for electron diffraction simulation.

	* The value of Pot_Type determines the barrier profile.
	
	input (with defolt values):
		 Fields of InputData argument and their default values:
			Dimensionality - 2
			Type - Type of kinetic operator discretisation  - 0 plane waves basis set - in this case
															- 1 sine basis set
			Pot_Type - type of the profile of the potential energy for barriers:
															- 20 free particle
															- 31 Infinite well for 1 slit
															- 32 Infinite well for 2 slits
															- 33 Infinite well for n slits 
			V0 - Potential energy height - 1
			mu - mass - 1
			Nx - Number of points along x-axis - 100
			Ny - Number of points along y-axis - 100
			x0 - Starting x-coordinate - 0
			y0 - Starting y-coordinate - 0
			xL - Ending x-coordinate - 1
			yL - Ending y-coordinate - 1

"""

OutData	= lambda:0
InputData = lambda:0

InputData.Type = 0
InputData.Dimensionality = 2

# ******************************************************************************#
# In calculations Type = 0 is used: plane waves basis set - the method of discretization of the kinetic energy operator

# The barrier profile:	:
# 20 --> free particle
# 31 --> Infinite well for 1 slit
# 32 --> Infinite well for 2 slits
# 33 --> Infinite well for n slits does not work.. Need to fix potential with n_slits
InputData.Pot_Type = 33

# Temporal used to make different calculations:
# Temporal==0 or Temporal==1 show only barrier profile
InputData.Temporal	= 1

InputData.mu		= 1		# The particle mass, a.u.: 1
InputData.V0		= 100 	# The height of the quantum barrier, a.u.: 100


InputData.Nx	= 100 	# The number of spatial points for the x-axis: 100
InputData.Ny	= 50	# The number of spatial points for the y-axis: 50

InputData.x0	= -50	# The left x-position of the calculation space, a.u.: -50
InputData.xL	=  50	# The right x-position of the calculation space, a.u.: 50
InputData.y0	= -50	# The left y-position of the calculation space, a.u.: -50
InputData.yL	=  50	# The right y-position of the calculation space, a.u.: 50

InputData.slit_d	= 3	# The slit width (changes in x-axis), a.u.: 8
InputData.slit_ls	= 8		# The slit length (changes in y-axis), a.u.: 4

#If the number of slits is more than one
InputData.slit_ly 	= 8		# The distance between centres of slits (period), a.u.: 4
InputData.n_slits = 5 # The number of slits: 3


# Parameters for the initial wave packet:
#-----------------------------------------------------------------------------
	## The initial position of the wave packet along
InputData.xc_WP = -25   # the x-axis: -25
InputData.yc_WP = 0	  # the y-axis: 0

	## The wave packet width in the direct space along
InputData.a0x	= 10	# the x-axis: 10
InputData.a0y	= 30	# the y-axis: 30

	## The wave vector of the wave packet center in the inverse direct space 
InputData.k0x	= 1		# of the x-axis: 1
InputData.k0y	= 0		# of the y-axis: 0

	## The number of points for the inverse space:
InputData.Nkx	= 25  # of the x-axis: 25
InputData.Nky	= 26  # of the y-axis: 26


# Parameters for temporal evolution of the wave packet :
#-----------------------------------------------------------------------------

	## Time parameters: 
InputData.Nt = 300  		# the number of points for time discretization: 300
InputData.delta_t = 0.2 	# The time step for simulations: 0.2
InputData.Nc_max 	= 10000	# Thr maximum number of iterations: 10000


InputData.amin = 5e-7 	# The accuracy of the Chebyshev decomposition: 5e-7
InputData.epsilon	= 1e-2	# The convergence criteria: 1e-2


	##The screen position for observing the diffraction: 20
InputData.x_screen = 20

#?
InputData.omegax	= 1
InputData.omegay	= 1

# Parameters for sep-like potentials: d = 5. sigma = 1e-2
InputData.d = 5.
InputData.sigma = 1e-2

InputData.isCheckOld = True
# InputData.isCheckOld = False

Lambda = 2 * np.pi/np.sqrt(InputData.k0x**2 + InputData.k0y**2)#max(k0x,k0y)
ls = 0
if(InputData.Pot_Type==31):
	ls = 2*Lambda
elif(InputData.Pot_Type==32 or InputData.Pot_Type==33):
	ls = Lambda/2
InputData.ls = ls

## main procedure
OutData,InputData = Get_Profile(InputData)
if hasattr(OutData,'Error'): # return error message if there is an error
	print (OutData)

Emax,Emin,pxmax,pymax,OutData.Eshift = Get_E_min_max2D(OutData.V,InputData.delta_x,InputData.delta_y,InputData.mu)
OutData.V -= OutData.Eshift


if(pxmax < InputData.k0x or pymax < InputData.k0y):
	OutData.Error	= 'Attention pxgrid < k0x or pygrid < k0y : increase Nx and Ny or decrease k0x and k0y'

#		sys.exit('Attention pxgrid<k0x or pygrid<k0y : increase Nx and Ny or decrease k0x and k0y') 

#   Build the grid for FT
OutData.kx,delta_kx,_,OutData.ky,delta_ky,_	= Get_Constant_K2D(InputData.Nkx,InputData.delta_x0,InputData.Nky,InputData.delta_y0)


#   Generate an intial wavepacket
Psi			= Get_Initial_Wave_Packet2D(OutData.x-InputData.xc_WP,OutData.y-InputData.yc_WP,InputData.k0x,InputData.a0x,InputData.k0y,InputData.a0y)
Psi_FT		= Get_Fourrier_Transform_WF2Dsingle(Psi,OutData.x,InputData.delta_x,OutData.y,InputData.delta_y,OutData.kx,OutData.ky)


if InputData.Temporal < 3: # show WP and barrier profile
	OutData.psi		= abs(Psi)**2
	OutData.psi_FT	= abs(Psi_FT)**2


Tx,Ty = Get_Kinetic_Operator2D(OutData.x,OutData.y,InputData.mu,InputData.LBOXx,InputData.LBOXy,InputData.Type)

H = Get_Hamiltonian_Operator2D(InputData.Nx,InputData.Ny,Tx,Ty,OutData.V)

time			= np.zeros(InputData.Nt, float)


#   Bessel coefficients for Chebyshev expansion
r = InputData.delta_t*(Emax - Emin)/2.
z = -ci*InputData.delta_t/r
Nc = int(max(2.0,r)) #nombre de paremtre Chebyshev de devloppement
a,Nc = Get_Chebyshev_Coefficients(Nc,InputData.Nc_max,InputData.amin,r)


OutData.Nc				= Nc
OutData.psi_t			= []
# OutData.psi_FT_t		= []
# OutData.psi_analytic_t	= []
time					= np.linspace(0,(InputData.Nt-1)*InputData.delta_t,InputData.Nt)


#    >>>>>>>>  the main time integration loop <<<<<<<<

# A = scipy.linalg.expm(-1j*InputData.delta_t*H)
for it in range(InputData.Nt):
	if it==0:
		Psi1 = np.dot(H,Psi.flatten()).reshape((InputData.Nx,InputData.Ny))
	else:
		Psi,Psi1	= Get_Propagator_Chebyshev_Expansion(Psi.flatten(),H,a,z)
		Psi			= Psi.reshape((InputData.Nx,InputData.Ny))
		Psi1		= Psi1.reshape((InputData.Nx,InputData.Ny))
		print(f"{it} / {InputData.Nt}")  # Pour voir la progression

	# if it > 0:
		# Psi = A @ Psi.flatten()

	OutData.psi_t.append(abs(Psi)**2)
	# OutData.psi_FT_t.append(abs(Psi_FT)**2)


	it += 1
	if it > InputData.Nt:
		it = InputData.Nt
	OutData.time				= time[:it]
```

### 2.3. Applications of Electron Interference: Cold Atom Gravimeter - Atomic Interferometry

<u>Objective:</u> To explore and demonstrate the applications of Cold Atom Gravimetry using Interferometry to measure gravitational acceleration and map gravitational fields with high precision. This study will illustrate how this method can be applied to geophysical surveys, structural analysis, and fundamental physics research by leveraging the wave-like properties of atoms to accurately detect variations in gravitational forces.

In this notebook, the application of atomic interferometry techniques will be presented. This phenomenon is used in the cold atom gravimeter, which allows for the measurement of gravitational acceleration $g$ with high precision, on the order of $2×10^{-8} g$ per second.

First, a few $10^8$ atoms are captured in a magneto-optical trap. They are cooled down to microKelvin temperatures before being released. During their free fall, they experience a sequence of three laser pulses, that split and recombine the atomic wavepackets. Finally, the phase difference between the two arms of the interferometer, that is proportionnal to g, is deduced form the measurement of the atomic state at the output of the interferometer.

More detailed information can be found in [[P. Cheinet](https://theses.hal.science/tel-00070861/document)], [[SYRTE](https://syrte.obspm.fr/spip/science/iaci/projets-en-cours/gravimetre/article/gravimetre-a-atomes-froids?lang=ent)] or [[Kavish Bhardwaj](https://arxiv.org/html/2405.04223v1)]

In Fig.6, the main components of the gravimeter setup are demonstrated. It consists of a magneto-optical trap (MOT) for $^{87}Rb$ atoms, a mechanism for the vertical launching of cold atoms in an atomic fountain, and Raman beams for the interferometric excitation of atoms within the fountain.

Two pairs of counter-propagating lasers induce Raman transitions with an effective wave vector $k_{eff}$. [[Z. Jiang](https://iopscience.iop.org/article/10.1088/0026-1394/49/6/666)].

Two counter-propagating laser beams, whose frequency difference resonates with the energy separation between the two hyperfine levels $∣a⟩ = ∣5S_{\frac{1}{2}}, F = 1⟩$ and $∣b⟩ = ∣5S_{\frac{1}{2}}, F = 2⟩$, induce a transition between these two levels (see, Fig.7). This transition is accompanied by a momentum transfer of $ \hbar(k_1 - k_2) = \hbar k_{eff}$ from the two beams, allowing for the spatial separation of the two states.

The following transitions are particularly distinguished:

- $\frac {\pi} 2$: an equal-probability splitting between the two states $|{a}\rangle$ and $|{b}\rangle$;

- $\pi$: a complete transfer from one state to the other;

After being selected in the state $|{a}\rangle$, the $^{87}Rb$ atoms will undergo, in order: a $\frac {\pi} 2$ transition (splitting), a $\pi$ transition (mirror), and a $\frac {\pi} 2$ transition (recombination).

The general principle of these three laser pulses is summarized in the diagram (Fig.7(a)).

</div>

To determine $g$, we need to know the probability of finding an atom in one of the states:

$$P=\frac {N_{|{a}\rangle}}{N_{|{a}\rangle}+ N_{|{b}\rangle}}= \frac 1 2 \left[ 1-\cos ( k_\text{eff}gT^2+\phi_1^0 - 2\phi_2^0 + \phi_3^0 )\right]=\frac 1 2 \left[ 1-\cos (\Delta \phi)\right]\tag{11}$$

where,

$g$ - Gravitational Acceleration.

$N_{|{a}\rangle}$ and $N_{|{b}\rangle}$ - The number of atoms detected in states $∣a⟩$ and $∣b⟩$, respectively, can be determined using the fluorescence method.

$k_{eff}$ - The effective wave vector, which represents the momentum transfer associated with the Raman beams in the atom interferometer. This parameter is a function of the wave vectors of the two counter-propagating laser beams and plays a critical role in the phase accumulation of the atoms.

$T$ - The time duration during which the atoms are allowed to fall freely between the initial state preparation and the measurement.

$\Delta \phi$ - The total phase difference experienced by the atoms due to gravitational acceleration and the interactions with the laser pulses. This phase shift is a result of both the gravitational influence on the freely falling atoms and the specific phases of the laser pulses, and it determines the final interference pattern observed in the atom interferometer.



In practice, a slightly different approach is used, involving the 'frequency sweep rate' $\alpha$. Due to the Doppler effect, the resonance frequency of the Raman process depends on velocity. As the atoms are in free fall, it is necessary to sweep the frequency difference between the Raman laser beams to satisfy the resonance condition for each pulse.

Thus, the phase shift $\alpha T^2$ is added to the gravitational phase shift (where $\alpha/2\pi$ is the frequency sweep rate). If $T$ is total time between the first and last laser pulses and  $\tau$ is the time between each consecutive pulse, the phase shift can be rewritten as:

$$\Delta \phi = (k_\text{eff}g-a)(T+2\tau)(T+4\tau/\pi)$$

The fringes are recorded by varying $\alpha$ for each measurement. When the Doppler effect is perfectly compensated, regardless of the value of $T$, the total phase shift $\Delta \phi$ becomes zero. The corresponding dark fringe associated with its sweep rate $\alpha_0$  allows to determine the value of $g$:

$$ g = \frac{\alpha_0}{k_{eff}}\tag{12}$$ 
​

### 2.4. Simulation of an electron in the gravitational field

```python
import numpy as np
from matplotlib.widgets import Button, Slider
from scipy import sparse as sp
from scipy import special
from scipy.sparse import linalg as lsp
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter

"""Système d'unité"""
hbar = 1  # Constante de Planck réduite
mu = 1  # Masse de l'électron
metre = 6.67e-9  # en unités naturelles (=Longueur d'un mètre dans notre système d'unité)
g = 6.67e-24  # en unités naturelles (= valeur de g dans notre système d'unité)


def Kinetic1D(N, Dx, mu=1):
    rx = (-1 / (2 * mu)) * (1 / Dx ** 2)  # Constante utile pour la résolution

    main_diag = np.full(N, -2 * rx)  # Main diagonal (central points)
    off_diag = np.full(N - 1, rx)  # Off-diagonal (neighboring points)

    # Create a sparse 1D derivative
    return sp.diags([off_diag, main_diag, off_diag], [-1, 0, 1], shape=(N, N))


def Hamiltonian2D(Dx, Dy, V, mu=1):
    Nx, Ny = V.shape

    # Create a sparse 1D Laplacian
    Tx = Kinetic1D(Nx, Dx, mu)
    Ty = Kinetic1D(Ny, Dy, mu)

    # Construct the 2D Laplacian using the Kronecker product of the 1D Laplacian with the identity matrix
    return sp.kron(Tx, np.eye(Ny)) + sp.kron(np.eye(Nx), Ty) + sp.diags(V.flatten(), offsets=0)


def Get_Propagator_Chebyshev_Expansion_1(Psi, Hz, a):
    """
    This fucntion advances Psi in time units of delta t by
    the Chebyshev expansion of the propagator.
    """
    g0, g1 = Psi, Hz @ Psi
    Psi = (float(a[0]) * g0) + (float(a[1]) * g1)

    for jt in range(2, len(a)):
        g0, g1 = g1, g0 + 2 * (Hz @ g1)  # reccurence of Chebyshev polynomials
        Psi += float(a[jt]) * g1  # accumulation step

    return Psi


def Get_Chebyshev_Coefficients(Nc, Nc_max, amin, r):
    a = np.zeros(Nc_max, float)
    C = 2
    a[0] = special.jv(0, r)
    k = np.arange(1, Nc)
    a[k] = C * special.jv(k, r)

    while a[Nc - 1] > amin:
        a[Nc] = C * special.jv(Nc, r)
        Nc += 1

        if Nc > Nc_max:
            print('The argument of the Chebyshev exp. is too large')
            break
    return a[:Nc], Nc


# Fonction principale, simulation dans boîte carrée
def simulation(psi0, V, Dx, Dy, Nt=200, mu=1, skip=5):
    """
    psi0: fonction d'onde initiale, taille Lx*Ly, (Nx-2)*(Ny-2) points
    V : potentiel, taille (NxN) points
    Dx: Pas spatial
    Dy: Pas spatial
    Nt: nombre d'itérations
    mu: Masse de l'électron
    """
    Nx, Ny = V.shape

    # Résolution du sysème Ax=My à chaque itération
    density = [np.abs(psi0) ** 2]  # Stockage de density0 dans la liste

    # Résolution via Cranck Nichoslon
    Dt = (Dx / 2) ** 2 + (Dy / 2) ** 2

    Emax = ((np.pi / Dx)**2 + (np.pi / Dy)**2) / (2 * mu) + np.max(V)
    Emin = np.min(V)
    Eshift = (Emax + Emin) / 2
    H = Hamiltonian2D(Dx, Dy, V - Eshift, mu)
    # H = Hamiltonian2D(Dx, Dy, V, mu)

    psi_vect = psi0.flatten()
    Energy0 = np.real(np.conj(psi_vect) @ H @ psi_vect) / np.real(np.conj(psi_vect) @ psi_vect)
    print(f"Initial WP energy: {Energy0:.5f} (potential min/max {np.min(V):.2f}/{np.max(V):.2f})")

    # E = sp.eye(H.shape[0], dtype=H.dtype)
    # Asp, Bsp = E + 1j * Dt / 2 * H, E - 1j * Dt / 2 * H


    #   Bessel coefficients for Chebyshev expansion
    Nc_max = 10000  # The maximum number of iterations
    amin = 5e-7  # The accuracy of the Chebyshev decomposition
    r = (skip*Dt) * (Emax - Emin) / 2
    z = -1j * (skip*Dt) / r
    Nc = int(max(2, r))  # nombre de paremtre Chebyshev de devloppement
    a, Nc = Get_Chebyshev_Coefficients(Nc, Nc_max, amin, r)
    Hz = H * z

    for i in range(1, Nt + 1):
        # psi_vect = lsp.spsolve(Asp, Bsp @ psi_vect)

        if i % skip == 0:
            psi_vect = Get_Propagator_Chebyshev_Expansion_1(psi_vect, Hz, a)
            print(f"{i} / {Nt}")  # Pour voir la progression
            # Ajout aux listes
            psi2 = np.abs(psi_vect.reshape((Nx, Ny))) ** 2
            density.append(psi2)

    print("\nSimulation complete")
    return density


# Paramètres de la simulation
Lx = 6
Ly = 6
Dx = 0.03
Dy = 0.03
Nx = int(Lx / Dx) + 1
Ny = int(Ly / Dy) + 1
# Nt = 1500
dt = 0.01
Nt = 50
skip = int(dt / ((Dx / 2) ** 2 + (Dy / 2) ** 2))  # save data after "skip" time points (skip = 1 - save all time points)
t_max = Nt * dt
Nt0 = Nt * skip

"""Fonction d'onde initiale"""
sigmax = 0.4  # Largeur du paquet d'onde initial
sigmay = 0.4  # Largeur du paquet d'onde initial
x0 = Lx / 2
y0 = Ly / 2

x = np.linspace(0, Lx, Nx)
y = np.linspace(0, Ly, Ny)
X, Y = np.meshgrid(x, y)

"""Potentiel"""
g = 50  # Valeur de g pour la simulation, arbitraire
V = g * Y

print(f"V shape = {V.shape}")
Energy0 = 1 / (2 * mu) * (1 / sigmax ** 2 + 1 / sigmay ** 2) / 2 + g * y0
print(f"Initial WP energy (analytical): {Energy0:.5f} (potential min/max {np.min(V):.2f}/{np.max(V):.2f})")

"""Simulation"""
psi0 = np.exp(-1 / 2 * (((X - x0) / sigmax) ** 2 + ((Y - y0) / sigmay) ** 2))  # Fonction d'onde initiale
psi0 /= np.real(np.sqrt(np.conj(psi0.flatten()) @ psi0.flatten()))
density = simulation(psi0, V, Dx, Dx, Nt0, mu, skip=skip)
print (density[int(t_max)][int(Nx/2)])

"""Visualisation de la fonction d'onde initiale et du potentiel"""
fig = plt.figure(figsize=(14, 6))
ax = (fig.add_subplot(121), fig.add_subplot(122, projection='3d'))  # Important: projection='3d'

# Densité de probabilité
cax1 = ax[0].imshow(density[0], extent=(0, Lx, 0, Ly), origin='lower', cmap='magma')
ax[0].set_title('Densité de probabilité')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
fig.colorbar(cax1, ax=ax[0])

# Potentiel
# cax2 = ax[1].imshow(V, extent=(0, Lx, 0, Ly), origin='lower', cmap='binary')
Norm = 0.1 * (np.max(V) - np.min(V)) / np.max(density[0])
cax2 = ax[1].plot_surface(X, Y, V, cmap='viridis', edgecolor='none')
Z = Energy0 + Norm * density[0]
cax3 = ax[1].plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
# cax3 = ax[1].plot_surface(X, Y, Z,
#                           facecolors=None,  # Transparent surface
#                           edgecolor=plt.cm.viridis(Z.flatten() / np.max(Z)),
#                           rstride=1, cstride=1)
ax[1].set_title('Potentiel')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
fig.colorbar(cax2, ax=ax[1])

plt.tight_layout()
# plt.show()


"""Animation"""
# Parameters for animation
pause = False
# Flag to prevent recursion during slider update
animation_is_playing = True

# fig = plt.figure(figsize=(9, 9))
# ax = fig.add_subplot(111, xlim=(0, Lx), ylim=(0, Ly))
# img = ax.imshow(density[0], extent=[0, Lx, 0, Ly], origin='lower', cmap="magma", vmin=0, zorder=1)  # Here the modulus of the 2D wave function shall be represented.
img = cax1
img2 = cax3


# Fonction pour l'animation
def animate(i):
    slider.set_val(i)  # Update slider position to match the current frame


# Function to toggle play/pause
def toggle_play(event):
    global pause, anim, animation_is_playing
    if pause:
        animation_is_playing = False
        anim.pause()  # Pause the animation
        play_button.label.set_text("►")
    else:
        animation_is_playing = True
        anim.resume()  # Pause the animation
        play_button.label.set_text("❚❚")
    pause = not pause


# Function to update frame with slider
def slider_update(val):
    global pause, anim, img2, animation_is_playing
    i = int(val)
    Z = density[i]
    img.set_data(Z)
    img.set_zorder(1)
    img.set_clim(vmin=0, vmax=np.max(Z))

    img2.remove()  # Remove the old surface
    Z = Energy0 + Norm * Z
    img2 = ax[1].plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
    # img2 = ax[1].plot_surface(X, Y, Energy0 + Norm * Z,
    #                           facecolors=None,  # Transparent surface
    #                           edgecolor=plt.cm.viridis(Z.flatten() / np.max(Z)),
    #                           rstride=1, cstride=1)

    fig.canvas.draw_idle()


pos = ax[0].get_position()
ax[0].set_position((pos.x0, pos.y0 + 0.1, pos.width, pos.height - 0.1))
ax_play = plt.axes([pos.x0, 0.05, 0.1 * (pos.x1 - pos.x0), 0.05])  # Play/Pause button position
ax_slider = plt.axes([pos.x0 + 0.15 * (pos.x1 - pos.x0), 0.05, 0.8 * (pos.x1 - pos.x0), 0.05])  # Slider position

play_button = Button(ax_play, "❚❚")  # Create Play/Pause button
play_button.on_clicked(toggle_play)

slider = Slider(ax_slider, "", 0, len(density) - 1, valinit=0, valstep=1)
slider.on_changed(slider_update)

anim = FuncAnimation(fig, animate, interval=100, frames=range(len(density)), blit=False, repeat=True)

# Sauvegarder l'animation en MP4
writer = FFMpegWriter(fps=30, metadata=dict(artist='Me'), bitrate=1800)
anim.save("pesanteur_electron_chebyshev_06.mp4", writer=writer)

plt.show()
```

$$ \rho = |e^{ik_yy + i\phi} + e^{-ik_yy}|^2$$

$$\rho = 4 \cos^2(k_y + \frac{\phi}{2})$$
