## 2. Interference of light: Double-slit Experiment

<u>Objective:</u>  To provide a foundational understanding of interference within its historical context, enabling a deeper comprehension of the fundamental nature of this phenomenon.

**Interference** (from Latin <i>interferens</i> is from <i>inter</i> - "between" and <i> ferre </i> - "to carry") is a phenomenon in which two coherent waves combine, taking into account their phase difference. This combination involves adding their displacements, resulting in a wave with either greater intensity (constructive interference) or reduced amplitude (destructive interference), depending on whether the waves are in phase or out of phase, respectively. 

- Interference effects can be observed with all types of waves, for example, light, radio, acoustic, surface water waves, gravity or matter waves.

- The result of interference (interference pattern) depends on the phase difference of the overlapping coherent waves (Fig.1).

Double-slit interference (Young's interference experiment), first conducted by Thomas Young in 1801, is a classic demonstration of the phenomenon of interference. This experiment shows that light behaves as a wave, producing interference patterns when light beams are superimposed.

The experiment involves passing a coherent light source (such as monochromatic light from a laser) through two closely spaced slits, thereby creating two coherent and in-phase light sources. The light from these slits propagates and overlaps on a screen placed behind the slits, producing a characteristic interference pattern of bright and dark fringes (Fig.1).

Let us consider a point $P$ on a screen located at a distance $L$ from two parallel slits illuminated by a coherent light wave with a wavelength $\lambda$. Coherence ensures that the light waves maintain a constant phase relationship, which is essential for producing a clear interference pattern.

Assuming that the light from the two sources, $S_1$ and $S_2$, is initially in phase, the phase relationship of the waves at point $P$ will depend on the optical path difference between the rays arriving from each slit: 

$$
 \Delta l= d \sin(\varphi)
$$
where, 
- $d$  est la distance entre les fentes;
- $\varphi$ est l'angle form√© par la ligne allant des fentes au point $ P $ avec l'axe central.

So, the <b>phase difference</b> between the rays coming from the two slits is then given by:

$$
 \Delta\phi = \frac{2\pi}{\lambda} \Delta l = \frac{2\pi d}{\lambda} \sin(\varphi) \tag{1}
$$

If the path difference is equal to an integral number of wavelengths the waves will interfere constructively (Fig.1(b)), leading to a bright spot on the screen.  Mathematically we can write this condition for maximum intensity as,

$$ d\sin\varphi = \pm m \lambda \tag{2}$$


Similarly, the condition for minimum intensity at P (destructive interference), when the path difference is a multiple of half wavelengths (waves will arrive out of phase), is given by

$$ d\sin\varphi = \pm (2m + 1)\frac{\lambda}{2}, \tag{3}$$

where $m = 1,2,3...$ order of diffraction. 



## 2.2. Intensity of light in Double-Slit

To determine how light intensity varies as a function of position on the screen in a double-slit experiment, one must consider the combined effects of <b>interference</b> from the coherent light waves emerging from the two slits and <b>diffraction</b> due to the finite width of each slit.


1. Interference. 

Based on previous descussions the electric fild in point $P$ can be written,

$$ E_p(\varphi) = E_{S_1} + E_{S_2} = E_0\left(\sin \omega t + \sin(\omega t + \Delta \phi)\right) = \\
= 2 E_0 \cos{\frac{\Delta \phi}{2}} \sin(\omega t + \frac{\Delta \phi}{2}).$$


So, intensity observed at $P$ is given by,

$$I_i(\varphi) = 4 E_0^2 \cos^2{\frac{\Delta \phi}{2}}$$


$$ I_i(\varphi) = I_0\cos^2\left(\frac{\pi d}{\lambda} \sin(\varphi) \right) \tag{4}$$

2. Diffraction.
The phenomenon when waves bend around obstacles or propagating through narrow openings, changing direction and deviating from geometric optics predictions, occurs due to the finite width of each slit.
The resulting diffraction pattern can be described using the sinc function, which accounts for the distribution of light due to the slit width:

$$ I_d(\varphi) = I_0  \text{sinc}^2\left(\frac{\pi a}{\lambda} \sin(\varphi)\right) \tag{5}$$

where $a$ is the width of the slit.

3. The overall light intensity $I(\varphi)$ observed at a point P on the screen is the product of the interference and diffraction contributions:

$$I(\varphi) = I_i(\varphi) I_d(\varphi)$$
$$I(\varphi) = I_0\cos^2\left(\frac{\pi d}{\lambda} \sin(\varphi) \right)\text{sinc}^2\left(\frac{\pi a}{\lambda} \sin(\varphi)\right) \tag{6}$$

Thus, the variation in light intensity on the screen in a double-slit experiment is governed by the interplay between interference and diffraction. Interference produces a series of bright and dark fringes, while diffraction influences the overall shape and visibility of these fringes, ultimately forming the observed intensity pattern. This relationship will be further explored and demonstrated in the next chapter.

###  2.2. Light Interference Modeling

Based on the Huygens-Fresnel principle, when light passes through a double-slit system, each point along the slits acts as a source of secondary spherical waves. If the slits have a width $a$ along the $x$-axis and their size in the other two directions is negligible (creating a simplified two-dimensional system), the electric field at an observation point \(x, y\) can be expressed as the superposition of the contributions from each point on the slits and reflects both the phase difference due to path length and the amplitude from each source, leading to the observed interference pattern.

$$
E(x,y) = \int_{x'}E_1(x,y)dx + \int_{x'}E_2(x, y)dx
$$

with

$$
E_1(x, y) = E_0 \frac{e^{ik\sqrt{(x - x')^2 + y^2}}}{\sqrt{(x - x')^2 + y^2}}
$$

$$
E_2(x, y) = E_0 \frac{e^{ik\sqrt{(x - x')^2 + y^2}}}{\sqrt{(x - x')^2 + y^2}}
$$

and $x' \in (-D/2+a,-D/2) \cup (D/2,D/2+a)$.

where 

$E_0$ is the amplitude of the waves passing through slits,

$D$ is the edge-to-edge distance between the two slits,

$k$ is the wavenumber (related to the wavelength $\lambda$ by $k = \frac{2\pi}{\lambda}$ ).

The intensity of spherical waves at a point \(x, y\) is given by the square modulus of the wave function:

$$
I(x, y) = |E(x, y)|^2 \tag{7}.
$$



```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Function to calculate and display wave intensity
def plot_wave_superposition(L=10, a=1.5, D=4, wavelength=1.5):
    
    Lx = 40 # Modeling area along the x-axis
    Lz = 30 # Modeling area along the z-axis
    
    k = 2 * np.pi / wavelength  # Wave Number


    # Creation of the Coordinate Grid
    x = np.linspace(0, Lx, 500)
    z = np.linspace(0, Lz, 500)
    X, Z = np.meshgrid(x, z, indexing="ij")

    # Positions of the Sources
    slits_x = np.hstack((Lx/2 + D/2 + np.linspace(0, a, 30),
                        Lx/2 - D/2 - np.linspace(0, a, 30)))

    # Initialize the superposition
    wave_superposition = 0

    for slit_x in slits_x:
        # Calculation of the Distances from the Grid Points to the Sources
        R = np.sqrt((X - slit_x) ** 2 + (Z - 0) ** 2)

        # Avoid division by zero by adding a small value to R
        wave = np.exp(1j * (k * R)) / (R + 1e-10)

        # Superposition of sinusoidal monochromatic waves
        wave_superposition += wave

    # Determine a fixed value of x to find the intensity distribution at thit position
    z_screen = L 
    index_z_screen = np.argmin(np.abs(z - z_screen))


    # Calculate intensity
    intensity = np.abs(wave_superposition)**2    
    # intensity = np.real(wave_superposition)


    # Normalize intensity for better visualization
    intensity = intensity / np.abs(intensity).max()

    # Calculate the intensity intensity on a screen for a fixed L.
    intensity_z_screen = intensity[:, index_z_screen]

    # Create a figure with two subplots
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7), gridspec_kw={'width_ratios': [2, 1]})

    # Visualization of wave superposition in all the slits
    im = ax1.imshow(intensity, extent=[0, Lz, 0, Lx], origin='lower', cmap='magma', vmax=intensity_z_screen.max()
    )
    ax1.set_title("Intensity")
    ax1.set_xlabel('z')
    ax1.set_ylabel('x')
    fig.colorbar(im, ax=ax1, label="Wave amplitude")

    # Visualization of the intensity on a screen for a fixed L.
    ax2.plot(x, intensity_z_screen, color='blue')
    ax2.set_title(f'Intensity at z = {z_screen}')
    ax2.set_xlabel('x')
    ax2.set_ylabel('Intensity')
  

# Utilisation de interact pour cr√©er les widgets de contr√¥le
interact(plot_wave_superposition,
            L=widgets.FloatSlider(min=5, max=50, step=1, value=10, description='L'),
            a=widgets.FloatSlider(min=0, max=4, step=0.1, value=1.5, description='a'),
            D=widgets.FloatSlider(min=0, max=10, step=0.1, value=1, description='D'),

            wavelength=widgets.FloatSlider(min=0.1, max=2, step=0.1, value=1.5, description='Wavelength, mkm'),
        )

```

```python
#calculation using analytical formulas
def diffraction_Franhover(k, slit_width, slit_height, x_screen, y_screen, z):
    X, Y =  np.meshgrid(x_screen, y_screen)
    beta_x = (k * slit_width / 2) * X/z + 1e-10
    beta_y = (k * slit_height / 2) * Y/z + 1e-10
    # Calculate the diffraction pattern for the single slit
    return (np.sin(beta_x) / beta_x) ** 2 * (np.sin(beta_y) / beta_y) ** 2

def interference_Franhover(k, D, slit_width, slit_height, x_screen, y_screen, z):
    X, Y =  np.meshgrid(x_screen, y_screen)
    # Calculate the interference pattern for the double slit
    alpha = (k * D / 2) * X / z
    interference_pattern = (np.cos(alpha)) ** 2
    return interference_pattern

def double_slit_Franhover(k, D, slit_width, slit_height, x_screen, y_screen, z):
    X, Y =  np.meshgrid(x_screen, y_screen)
    beta_x = (k * slit_width / 2) * X/z + 1e-10
    beta_y = (k * slit_height / 2) * Y/z + 1e-10
    single_slit_diffraction = (np.sin(beta_x) / beta_x) ** 2 * (np.sin(beta_y) / beta_y) ** 2
    
    # Calculate the interference pattern for the double slit
    alpha = (k * D / 2) * X / z
    interference_pattern = (np.cos(alpha)) ** 2
    
    # Combine the single-slit diffraction pattern with the interference pattern
    return single_slit_diffraction * interference_pattern


#numerical calculation: using Fast Fourier Transform (eq.11)
def diffraction_FFT(k, sheet, z):
    intensity = fftshift(fft2(sheet.mask * np.exp(1j * k/(2*z) * (sheet.xx**2 + sheet.yy**2))))
    intensity = np.absolute(intensity)**2
    return intensity / intensity.max()



def get_intensity(field):
    # Calculate the intensity (square of the field magnitude)
    intensity = np.abs(field)**2

    # Normalize the intensity
    return intensity / intensity.max()

#direct integtation (eq.8)
def diffraction_direct(k, sheet, x_screen, y_screen, z):
    N_x, N_y = x_screen.size, y_screen.size

    # Initialize the field on the screen
    field = np.zeros((N_y, N_x), dtype=np.complex128)

    # Define the integrand function
    def integrand(k, slit_x, slit_y, x, y, z):
        """
        integrand function consisting of spherical wave
        """
        # Calculate the distance from the slit point to the screen point
        kr = k * np.sqrt((x - slit_x)**2 + (y - slit_y)**2 + z**2)
        return np.exp(1j * kr) / kr

    # Loop over each point in the slit and calculate its contribution to the screen
    for ix in range(N_x):
        for iy in range(N_y):
            # Calculate the contribution to the field at the screen point
            field[iy][ix] = integrand(k, sheet.xx, sheet.yy, x_screen[ix], y_screen[iy], z).sum()

    return field
```

### 2.3. Light Interference Applications

Light interference has numerous applications across various fields, ranging from scientific research to practical technologies. One of the main components is interferometr - devices that extract information from interference. The Michelson Interferometer is one of the most common types of interferometers. The setup, presented in Fig.2(a) consists of a semi-reflective mirror, called a beam splitter, which divides a beam into two perpendicular beams. Beam 1 reflects off Mirror 1, then crosses the beam splitter again to meet the photodetector (PD). Beam 2 passes through the beam splitter, reflects off Mirror 2, and is then reflected by the beam splitter before reaching the photodetector. By introducing different path lengths or materials in each beam‚Äôs path, a phase shift is created between the two optical paths.

When the path difference corresponds to a whole number of wavelengths (0, $\lambda$, $2\lambda$‚Ä¶), constructive interference occurs, creating a strong signal. If the difference is an odd multiple of half wavelengths ($\lambda/2$, $3\lambda/2$‚Ä¶), destructive interference occurs, resulting in a near-zero signal at the photodetector.

The distances between the beam splitter and the mirrors are called arms. One of the mirrors (M1) can be rotated and the other can be mover along the system's axis. This allows adjustments to both the path difference between beams and the direction of their propagation. Depending on the mirror positions, different interference patterns can be observed. If two surfaces are strictly perpendicular to each other, the interference pattern will appear as fringes of equal inclination (the system of concentric circles). If the mirrors are not perfectly perpendicular, the pattern takes the form of fringes of equal thickness (the parallel stripes system).

## 3. Matter Interference

The double-slit experiment demonstrates that particles, such as electrons or photons, passing through two slits create an interference pattern on a screen, proving that they behave as waves. This experiment in modern physics is a key demonstration of wave-particle duality - a fundamental concept in quantum mechanics. The hypothesis of the dual nature of waves and matter was first proposed by De Broglie in 1924, suggesting that each particle with momentum $p$ has a corresponding wavelength $\lambda$, related to its momentum by the same equation as for photons:

$$\lambda = \frac{h}{p}\tag{8}$$

where $h$ is Planck's constant.

This type of experiment was first performed by Thomas Young in 1801, as a demonstration of the wave behavior of visible light. In 1927, Davisson and Germer demonstrated that electrons exhibit the same behavior, which was later extended to atoms and molecules, thus confirming de Broglie's hypothesis.




### 3.1. The Electron Interference

Young's double-slit experiment, initially conducted with light, has been revisited and extended to various particles, most famously electrons. In 1961, Claus J√∂nsson performed the first successful double-slit experiment with electrons. In his setup, electrons were accelerated from a source to an energy of 50 keV. The slits, which were extremely narrow (about 0.5 Œºm) and spaced approximately 0.1 Œºm apart, were crafted on a thin copper foil. ### 3.2. Simulation of matter waves

<u>Objective:</u> To develop a deeper understanding of the underlying physics of electron diffraction by simulating electron diffraction patterns and analyzing the effects of various parameters on the temporal evolution of electrons in space.

The temporal evolution of electrons in space is described of time-dependent Schrodinger equation:
$$
i\hbar {\frac {d}{dt}} \Psi ={\hat {H}} \Psi, \tag{9}
$$
The solution of Eq.(9) can be written in operator form:
$$
\Psi(t) = e^{ -\frac{i}{\hbar} {\hat {H}} t} \Psi(t=0), \tag{10}
$$
where $\Psi(t)$ is the time evolution of the wave function, starting from the initial state $\Psi(t=0)$.

The exponential operator in Eq.(10) can be expanded in terms of Chebyshev polynomials. The Chebyshev expansion method is often used to approximate functions of operators, especially for large-scale problems, due to its numerical efficiency and stability. The core idea is to expand the exponential operator as a series of Chebyshev polynomials $T_n(x)$, where $x$ is a rescaled version of the Hamiltonian $\hat{H} / \hbar$, ensuring that the spectrum of $x$ lies in the interval $[-1, 1]$, which is the domain of Chebyshev polynomials.

The function $f(x) = e^{-i x t}$ can be expanded in terms of Chebyshev polynomials $T_n(x)$:
$$
   f(x) = e^{-i x t} \approx \sum_{n=0}^{N} c_n T_n(x),
$$

where $T_n(x)$ are Chebyshev polynomials of the first kind, and the coefficients $c_n$ are given by:
$$
   c_n = \frac{2}{\pi} \int_{-1}^{1} \frac{e^{-i t x} T_n(x)}{\sqrt{1 - x^2}} dx = i^n J_n(t),
$$
where $J_n(t)$ are the Bessel functions of the first kind.

Once the expansion coefficients $c_n$ are determined, the time-evolved state can be computed as a linear combination of Chebyshev polynomials applied to the rescaled Hamiltonian $\hat{H}$:
$$
   \Psi(t) \approx \sum_{n=0}^{N} c_n T_n(\hat{H}) \Psi(0).
$$
   
The action of the Chebyshev polynomials $T_n(\hat{H})$ on the initial state $\Psi(0)$ is computed recursively using the recurrence relation for Chebyshev polynomials:
   
$$
   T_0(x) = 1, \quad T_1(x) = x, \quad T_{n+1}(x) = 2x T_n(x) - T_{n-1}(x).
$$
   
So, starting with $\Psi_0 = \Psi(t=0)$ and $\Psi_1 = \hat{H} \Psi_0$, we can compute $\Psi_n = T_n(\hat{H}) \Psi_0$ recursively for $n = 2, 3, \dots, N$.

In practice, the series is truncated at a finite number $N$, which is determined by the desired accuracy and the properties of the Hamiltonian and time evolution. The error decreases exponentially with the number of terms due to the rapid convergence of the Chebyshev expansion for smooth functions.

The initial state $\Psi(0)$ can be presentred int the form Gaussian wave packet $\Psi(t=0) \propto \exp\left({-(x/x_0)^2 - (y/y_0)^2}\right)$.

```python
def Get_Propagator_Chebyshev_Expansion(Psi,H,a,z):
	"""
		This fucntion advances Psi in time units of delta t by 
		the Chebyshev expansion of the propagator.
	
	"""
	Hz = H*z
	g0,g1 = Psi, np.dot(Hz,Psi)
	Psi = a[0]*g0 + a[1]*g1

	for jt in range(2,len(a)):
		g0,g1 = g1, g0 + np.dot(Hz,g1)*2 # reccurence of Chebyshev polynomials
		Psi += a[jt]*g1 # accumulation step

	Psi1 = np.dot(H,Psi)
	return Psi,Psi1

def Get_Chebyshev_Coefficients(Nc,Nc_max,amin,r):
	a=np.zeros(Nc_max,float)
	C=2
	a[0] = special.jv(0,r)
	k = np.arange(1,Nc)
	a[k] = C*special.jv(k,r)

	while (a[Nc-1] > amin):
		a[Nc] = C*special.jv(Nc,r)
		Nc+=1

		if(Nc>Nc_max):
			print('The argument of the Chebyshev exp. is too large')
			break
	return a[:Nc],Nc


def Get_Initial_Wave_Packet2D(x,y,k0x,a0x,k0y,a0y):
	Ax = (2/(np.pi*a0x**2))**(1/4)*np.exp(-(x/a0x)**2)*np.exp(ci*k0x*x)
	Ay = (2/(np.pi*a0y**2))**(1/4)*np.exp(-(y/a0y)**2)*np.exp(ci*k0y*y)
	Psi=	np.einsum( 'x,y->xy',Ax,Ay )
	return Psi

# FFT for single eigen state ( psi(x,y) )
def Get_Fourrier_Transform_WF2Dsingle(psi,x,delta_x,y,delta_y,kx,ky):
	KX,X = np.meshgrid(kx,x, sparse=False, indexing='ij')
	KY,Y = np.meshgrid(ky,y, sparse=False, indexing='ij')

	psi_FT = ( (np.exp(-ci*(KX*X)) * delta_x) @ psi @  (np.exp(-ci*(KY*Y))*delta_y).T )/(2*np.pi) # Integrate Trapeze
	return psi_FT

def Get_Kinetic_Operator(x,mu,LBOXx,Type):
	""" 
		Calculation of the kinetic operator

		inputs:
			x -	vector of  coordinate
			mu -mass
			LBOXx- the length of simulation box
			Type - type of the discretization for the kinetic operator

		Outputs:
			T - 2D array of the kinetic operator 
	
	"""
	N = len(x)
	if(Type==0):
		coef= (np.pi/LBOXx)**2/mu
		Di	= np.arange(1,N)
		if(N%2==0):
			T_diag		= (N*N+2)/6.
			T_nondiag	= (-1.)**Di/(np.sin(Di*np.pi/N))**2
			#print('<br>even T=',coef*hstack( (T_diag, T_nondiag) ))
		else:
			T_diag		= (N*N-1)/6.
			T_nondiag	= (-1.)**Di*np.cos(Di*np.pi/N)/(np.sin(Di*np.pi/N))**2
		T = toeplitz( coef*np.hstack( (T_diag, T_nondiag) ) )
	elif(Type==1):
		# np.sin basis set
		coef	= (np.pi/LBOXx)**2/(2*mu)/2
		T = np.zeros((N,N),float)
		I1, I2 = np.meshgrid(np.arange(N),np.arange(N))
		cind = np.logical_not(I1==I2)
		DI = I1[cind]-I2[cind]
		SI = I1[cind]+I2[cind]
		T[cind] = coef*(-1.)**DI*(1./np.sin(np.pi*DI/(2*(N+1)))**2 - 1./np.sin(np.pi*(SI+2)/(2*(N+1)))**2)
		cind = (I1==I2)
		T[cind] = coef*((2*(N+1)**2+1)/3. - 1/np.sin(np.pi*(I1[cind]+1)/(N+1))**2)
	return T


def Get_Kinetic_Operator2D(x,y,mu,LBOXx,LBOXy,Type=0):
	return Get_Kinetic_Operator(x,mu,LBOXx,Type), Get_Kinetic_Operator(y,mu,LBOXy,Type)


def Get_Hamiltonian_Operator2D(Nx,Ny,Tx,Ty,V):

	return   (np.einsum('ij,kl->ikjl', Tx, np.eye(Ny,Ny))+\
			np.einsum('ij,kl->ljki', Ty, np.eye(Nx,Nx))).reshape((Nx*Ny,Nx*Ny))+\
			np.diag(V.flatten()) # H[xy,xy]


def Get_E_min_max2D(V,delta_x,delta_y,mu):
	pxmax = np.pi/delta_x.min()
	pymax = np.pi/delta_y.min()
	Emax = (pxmax**2+pymax**2)/(2.*mu)+V.max()
	Emin = V.min()
	Eshift = (Emax+Emin)/2

	return Emax,Emin,pxmax,pymax,Eshift

def Get_Constant_K(Nk,delta_x):
	kmax = np.pi/delta_x.max()
	k = np.linspace(-kmax,kmax+kmax/Nk, num=2*Nk+1, endpoint=False)
	dx = np.diff(k)
	delta_k = (np.concatenate([[0],dx]) + np.concatenate([dx,[0]]))/2										
	return k,delta_k,kmax

def Get_Constant_K2D(Nkx,delta_x,Nky,delta_y):
	kx,delta_kx,kxmax = Get_Constant_K(Nkx,delta_x)
	ky,delta_ky,kymax = Get_Constant_K(Nky,delta_y)

	return kx,delta_kx,kxmax,ky,delta_ky,kymax
```

```python
def Get_Profile(InputData):
	"""
		Calculate potential energy profile. 

		input:
		 Fields of InputData
		
		output:
		 Fields of OutData argument :
			V - Vector of the potential energy data
			x - x-coordinate
			y - y-coordinate
	"""
	OutData				= lambda:0

	InputData.xc = (InputData.xL+InputData.x0)/2
	InputData.yc = (InputData.yL+InputData.y0)/2
	
	# Profile type

	if(InputData.Pot_Type==20):
		InputData.xmin_g,InputData.xmax_g	= InputData.x0,InputData.xL
		InputData.ymin_g,InputData.ymax_g	= InputData.y0,InputData.yL
		
		OutData.x,InputData.delta_x0,InputData.delta_x,OutData.y,InputData.delta_y0,InputData.delta_y,InputData.LBOXx,InputData.LBOXy	= Get_Constant_Grid2D(InputData.Nx,InputData.x0,InputData.xL,InputData.Ny,InputData.y0,InputData.yL,InputData.Type)
			
		#   Build the potential 
		OutData.V = Get_Potential_Operator2D(OutData.x,OutData.y,None,InputData.Pot_Type)

	elif(InputData.Pot_Type==31): # 1 slit	
		InputData.xmin_g,InputData.xmax_g	= InputData.x0,InputData.xL
		InputData.ymin_g,InputData.ymax_g	= InputData.y0,InputData.yL

		OutData.x,InputData.delta_x0,InputData.delta_x,OutData.y,InputData.delta_y0,InputData.delta_y,InputData.LBOXx,InputData.LBOXy	= Get_Constant_Grid2D(InputData.Nx,InputData.x0,InputData.xL,InputData.Ny,InputData.y0,InputData.yL,InputData.Type)
		
		#  Build the potential 
		OutData.V = Get_Potential_Operator2D(OutData.x,OutData.y,{'V0':InputData.V0,'l':InputData.slit_d,'ls':InputData.slit_ls,'xc':InputData.xc,'yc':InputData.yc},InputData.Pot_Type)

	elif(InputData.Pot_Type==32): # 2 slits
		InputData.xmin_g,InputData.xmax_g	= InputData.x0,InputData.xL
		InputData.ymin_g,InputData.ymax_g	= InputData.y0,InputData.yL

		OutData.x,InputData.delta_x0,InputData.delta_x,OutData.y,InputData.delta_y0,InputData.delta_y,InputData.LBOXx,InputData.LBOXy	= Get_Constant_Grid2D(InputData.Nx,InputData.x0,InputData.xL,InputData.Ny,InputData.y0,InputData.yL,InputData.Type)
		
		#  Build the potential 
		OutData.V = Get_Potential_Operator2D(OutData.x,OutData.y,{'V0':InputData.V0,'l':InputData.slit_d,'ls':InputData.slit_ls,'ly':InputData.slit_ly,'xc':InputData.xc,'yc':InputData.yc},InputData.Pot_Type)

	elif(InputData.Pot_Type==33): # N slits
		if(InputData.n_slits * InputData.slit_ls + (InputData.n_slits - 1+2) * InputData.slit_ly >= (InputData.yL - InputData.y0)):
			sys.exit('Number of slits too large for the current grid!!!') 

		print("Get_Profile",InputData.n_slits)

		InputData.xmin_g,InputData.xmax_g	= InputData.x0,InputData.xL
		InputData.ymin_g,InputData.ymax_g	= InputData.y0,InputData.yL

		OutData.x,InputData.delta_x0,InputData.delta_x,OutData.y,InputData.delta_y0,InputData.delta_y,InputData.LBOXx,InputData.LBOXy	= Get_Constant_Grid2D(InputData.Nx,InputData.x0,InputData.xL,InputData.Ny,InputData.y0,InputData.yL,InputData.Type)

		#  Build the potential 
		OutData.V = Get_Potential_Operator2D(OutData.x,OutData.y,{'V0':InputData.V0,'l':InputData.slit_d,'ls':InputData.slit_ls,'ly':InputData.slit_ly,'n_slits':InputData.n_slits,'xc':InputData.xc,'yc':InputData.yc},InputData.Pot_Type)

	else:
		OutputData.Error = 'The type of potential energy profile is not correct (Pot_Type = %i)!'%(InputData.Pot_Type) # type: ignore

	return OutData,InputData
```

```python
"""
	Etrer your data for electron diffraction simulation.

	* The value of Pot_Type determines the barrier profile.
	
	input (with defolt values):
		 Fields of InputData argument and their default values:
			Dimensionality - 2
			Type - Type of kinetic operator discretisation  - 0 plane waves basis set - in this case
															- 1 sine basis set
			Pot_Type - type of the profile of the potential energy for barriers:
															- 20 free particle
															- 31 Infinite well for 1 slit
															- 32 Infinite well for 2 slits
															- 33 Infinite well for n slits 
			V0 - Potential energy height - 1
			mu - mass - 1
			Nx - Number of points along x-axis - 100
			Ny - Number of points along y-axis - 100
			x0 - Starting x-coordinate - 0
			y0 - Starting y-coordinate - 0
			xL - Ending x-coordinate - 1
			yL - Ending y-coordinate - 1

"""

OutData	= lambda:0
InputData = lambda:0

InputData.Type = 0
InputData.Dimensionality = 2

# ******************************************************************************#
# In calculations Type = 0 is used: plane waves basis set - the method of discretization of the kinetic energy operator

# The barrier profile:	:
# 20 --> free particle
# 31 --> Infinite well for 1 slit
# 32 --> Infinite well for 2 slits
# 33 --> Infinite well for n slits does not work.. Need to fix potential with n_slits
InputData.Pot_Type = 32

# Temporal used to make different calculations:
# Temporal==0 or Temporal==1 show only barrier profile
InputData.Temporal	= 1

InputData.mu		= 1		# The particle mass, a.u.: 1
InputData.V0		= 100 	# The height of the quantum barrier, a.u.: 100


InputData.Nx	= 100 	# The number of spatial points for the x-axis: 100
InputData.Ny	= 50	# The number of spatial points for the y-axis: 50

InputData.x0	= -50	# The left x-position of the calculation space, a.u.: -50
InputData.xL	=  50	# The right x-position of the calculation space, a.u.: 50
InputData.y0	= -50	# The left y-position of the calculation space, a.u.: -50
InputData.yL	=  50	# The right y-position of the calculation space, a.u.: 50

InputData.slit_d	= 3	# The slit width (changes in x-axis), a.u.: 8
InputData.slit_ls	= 8		# The slit length (changes in y-axis), a.u.: 4

#If the number of slits is more than one
InputData.slit_ly 	= 8		# The distance between centres of slits (period), a.u.: 4
InputData.n_slits = 5 # The number of slits: 3


# Parameters for the initial wave packet:
#-----------------------------------------------------------------------------
	## The initial position of the wave packet along
InputData.xc_WP = -25   # the x-axis: -25
InputData.yc_WP = 0	  # the y-axis: 0

	## The wave packet width in the direct space along
InputData.a0x	= 10	# the x-axis: 10
InputData.a0y	= 30	# the y-axis: 30

	## The wave vector of the wave packet center in the inverse direct space 
InputData.k0x	= 1		# of the x-axis: 1
InputData.k0y	= 0		# of the y-axis: 0

	## The number of points for the inverse space:
InputData.Nkx	= 25  # of the x-axis: 25
InputData.Nky	= 26  # of the y-axis: 26


# Parameters for temporal evolution of the wave packet :
#-----------------------------------------------------------------------------

	## Time parameters: 
InputData.Nt = 300  		# the number of points for time discretization: 300
InputData.delta_t = 0.2 	# The time step for simulations: 0.2
InputData.Nc_max 	= 10000	# Thr maximum number of iterations: 10000


InputData.amin = 5e-7 	# The accuracy of the Chebyshev decomposition: 5e-7
InputData.epsilon	= 1e-2	# The convergence criteria: 1e-2


	##The screen position for observing the diffraction: 20
InputData.x_screen = 20

#?
InputData.omegax	= 1
InputData.omegay	= 1

# Parameters for sep-like potentials: d = 5. sigma = 1e-2
InputData.d = 5.
InputData.sigma = 1e-2

InputData.isCheckOld = True
# InputData.isCheckOld = False

Lambda = 2 * np.pi/np.sqrt(InputData.k0x**2 + InputData.k0y**2)#max(k0x,k0y)
ls = 0
if(InputData.Pot_Type==31):
	ls = 2*Lambda
elif(InputData.Pot_Type==32 or InputData.Pot_Type==33):
	ls = Lambda/2
InputData.ls = ls
print("x-coordinate for which the diffraction spot could be observed is ls**2/Lambda=%g"%(InputData.ls**2/Lambda + (InputData.xL + InputData.x0)/2.))
print("y coordinate for which the diffraction spot could be observed is ls**2/Lambda=%g"%(InputData.ls**2/Lambda))


## main procedure
OutData,InputData = Get_Profile(InputData)
if hasattr(OutData,'Error'): # return error message if there is an error
	print (OutData)

Emax,Emin,pxmax,pymax,OutData.Eshift = Get_E_min_max2D(OutData.V,InputData.delta_x,InputData.delta_y,InputData.mu)
OutData.V -= OutData.Eshift


if(pxmax < InputData.k0x or pymax < InputData.k0y):
	OutData.Error	= 'Attention pxgrid < k0x or pygrid < k0y : increase Nx and Ny or decrease k0x and k0y'

#		sys.exit('Attention pxgrid<k0x or pygrid<k0y : increase Nx and Ny or decrease k0x and k0y') 

#   Build the grid for FT
OutData.kx,delta_kx,_,OutData.ky,delta_ky,_	= Get_Constant_K2D(InputData.Nkx,InputData.delta_x0,InputData.Nky,InputData.delta_y0)


#   Generate an intial wavepacket
Psi			= Get_Initial_Wave_Packet2D(OutData.x-InputData.xc_WP,OutData.y-InputData.yc_WP,InputData.k0x,InputData.a0x,InputData.k0y,InputData.a0y)
Psi_FT		= Get_Fourrier_Transform_WF2Dsingle(Psi,OutData.x,InputData.delta_x,OutData.y,InputData.delta_y,OutData.kx,OutData.ky)


if InputData.Temporal < 3: # show WP and barrier profile
	OutData.psi		= abs(Psi)**2
	OutData.psi_FT	= abs(Psi_FT)**2


Tx,Ty = Get_Kinetic_Operator2D(OutData.x,OutData.y,InputData.mu,InputData.LBOXx,InputData.LBOXy,InputData.Type)

H = Get_Hamiltonian_Operator2D(InputData.Nx,InputData.Ny,Tx,Ty,OutData.V)

norm			= np.zeros(InputData.Nt, float)
H_average		= np.zeros(InputData.Nt, float)
Delta_X			= np.zeros(InputData.Nt, float)
Delta_Y			= np.zeros(InputData.Nt, float)
Delta_Px		= np.zeros(InputData.Nt, float)
Delta_Py		= np.zeros(InputData.Nt, float)
Delta_X_analytic= np.zeros(InputData.Nt, float)
Delta_Y_analytic= np.zeros(InputData.Nt, float)
x_average		= np.zeros(InputData.Nt, float)
y_average		= np.zeros(InputData.Nt, float)
px_average		= np.zeros(InputData.Nt, float)
py_average		= np.zeros(InputData.Nt, float)
Heisenberg		= np.zeros(InputData.Nt, float)
Heisenberg_X	= np.zeros(InputData.Nt, float)
Heisenberg_Y	= np.zeros(InputData.Nt, float)
time			= np.zeros(InputData.Nt, float)


#   Bessel coefficients for Chebyshev expansion
r = InputData.delta_t*(Emax - Emin)/2.
z = -ci*InputData.delta_t/r
Nc = int(max(2.0,r)) #nombre de paremtre Chebyshev de devloppement
a,Nc = Get_Chebyshev_Coefficients(Nc,InputData.Nc_max,InputData.amin,r)


OutData.Nc				= Nc
OutData.psi_t			= []
OutData.psi_FT_t		= []
OutData.psi_analytic_t	= []
time					= np.linspace(0,(InputData.Nt-1)*InputData.delta_t,InputData.Nt)

#    >>>>>>>>  the main time integration loop <<<<<<<<
for it in range(InputData.Nt):
	if it==0:
		Psi1 = np.dot(H,Psi.flatten()).reshape((InputData.Nx,InputData.Ny))
	else:
		Psi,Psi1	= Get_Propagator_Chebyshev_Expansion(Psi.flatten(),H,a,z)
		Psi			= Psi.reshape((InputData.Nx,InputData.Ny))
		Psi1		= Psi1.reshape((InputData.Nx,InputData.Ny))

	OutData.psi_t.append(abs(Psi)**2)
	OutData.psi_FT_t.append(abs(Psi_FT)**2)


	it += 1
	if it > InputData.Nt:
		it = InputData.Nt
	OutData.time				= time[:it]
	OutData.H_average			= H_average[:it]
	OutData.Delta_X_analytic	= Delta_X_analytic[:it]
	OutData.Delta_Y_analytic	= Delta_Y_analytic[:it]
	OutData.norm				= norm[:it]
	OutData.Heisenberg			= Heisenberg[:it]
	OutData.Delta_X				= Delta_X[:it]
	OutData.Delta_Y				= Delta_Y[:it]
	OutData.Delta_Px			= Delta_Px[:it]
	OutData.Delta_Py			= Delta_Py[:it]
	OutData.x_average			= x_average[:it]
	OutData.y_average			= y_average[:it]
	OutData.px_average			= px_average[:it]
	OutData.py_average			= py_average[:it]
	OutData.Heisenberg			= Heisenberg[:it]
	OutData.Heisenberg_X		= Heisenberg_X[:it]
	OutData.Heisenberg_Y		= Heisenberg_Y[:it]
```

### 3.3. Applications of Electron Interference: Cold Atom Gravimeter - Atomic Interferometry

<u>Objective:</u> To explore and demonstrate the applications of Cold Atom Gravimetry using Interferometry to measure gravitational acceleration and map gravitational fields with high precision. This study will illustrate how this method can be applied to geophysical surveys, structural analysis, and fundamental physics research by leveraging the wave-like properties of atoms to accurately detect variations in gravitational forces.

In this notebook, the application of atomic interferometry will be presented. This phenomenon is used in the cold atom gravimeter, which allows for the measurement of gravitational acceleration $g$ with high precision, on the order of $2√ó10^{-8} g$ per second.

The absolute gravimeter is based on atom interferometry techniques applied to cold atoms. First, a few $10^8$ atoms are captured in a magneto-optical trap. They are cooled down to microKelvin temperatures before being released. During their free fall, they experience a sequence of three laser pulses, that split and recombine the atomic wavepackets. Finally, the phase difference between the two arms of the interferometer, that is proportionnal to g, is deduced form the measurement of the atomic state at the output of the interferometer.
Here, we will provide a description of the essential points of its operation. 

More detailed information can be found in [[P. Cheinet](https://theses.hal.science/tel-00070861/document)], [[SYRTE](https://syrte.obspm.fr/spip/science/iaci/projets-en-cours/gravimetre/article/gravimetre-a-atomes-froids?lang=ent)] or [[Kavish Bhardwaj](https://arxiv.org/html/2405.04223v1)]

In Fig.6, the main components of the gravimeter setup are demonstrated. It consists of a magneto-optical trap (MOT) for $^{87}Rb$ atoms, a mechanism for the vertical launching of cold atoms in an atomic fountain, and Raman beams for the interferometric excitation of atoms within the fountain.

To determine $g$, we need to know the probability of finding an atom in one of the states:

$$P=\frac {N_{|{a}\rangle}}{N_{|{a}\rangle}+ N_{|{b}\rangle}}= \frac 1 2 \left[ 1-\cos ( k_\text{eff}gT^2+\phi_1^0 - 2\phi_2^0 + \phi_3^0 )\right]=\frac 1 2 \left[ 1-\cos (\Delta \phi)\right]\tag{11}$$

where,

$g$: Gravitational Acceleration.

$N_{|{a}\rangle}$ and $N_{|{b}\rangle}$: The number of atoms detected in states $‚à£a‚ü©$ and $‚à£b‚ü©$, respectively, can be determined using the fluorescence method."

$k_{eff}$: The effective wave vector, which represents the momentum transfer associated with the Raman beams in the atom interferometer. This parameter is a function of the wave vectors of the two counter-propagating laser beams and plays a critical role in the phase accumulation of the atoms.

$T$: The time duration during which the atoms are allowed to fall freely between the initial state preparation and the measurement.

$\Delta \phi$: The total phase difference experienced by the atoms due to gravitational acceleration and the interactions with the laser pulses. This phase shift is a result of both the gravitational influence on the freely falling atoms and the specific phases of the laser pulses, and it determines the final interference pattern observed in the atom interferometer.



In practice, a slightly different approach is used, involving the 'frequency sweep rate' $\alpha$. Due to the Doppler effect, the resonance frequency of the Raman process depends on velocity. As the atoms are in free fall, it is necessary to sweep the frequency difference between the Raman laser beams to satisfy the resonance condition for each pulse.

Thus, the phase shift $\alpha T^2$ is added to the gravitational phase shift (where $\alpha/2\pi$ is the frequency sweep rate). If $T$ is total time between the first and last laser pulses and  $\tau$ is the time between each consecutive pulse, the phase shift can be rewritten as:

$$\Delta \phi = (k_\text{eff}g-a)(T+2\tau)(T+4\tau/\pi)$$

The fringes are recorded by varying $\alpha$ for each measurement. When the Doppler effect is perfectly compensated, regardless of the value of $T$, the total phase shift $\Delta \phi$ becomes zero. The corresponding dark fringe associated with its sweep rate $\alpha_0$  allows to determine the value of $g$:

$$ g = \frac{\alpha_0}{k_{eff}}\tag{12}$$
