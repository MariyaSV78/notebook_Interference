# Les ondes électromagnétiques, diffraction et interférences

En mécanique ondulatoire, les interférences résultent de la combinaison de deux ondes capables d'interagir. Lorsque deux ondes électromagnétiques de même nature se rencontrent, elle se superposent. Cette superposition peut être constructive, où les amplitudes des ondes s'additionnent, créant ainsi une onde de plus grande amplitude, ou destructive, où les amplitudes se soustraient, réduisant ou annulant l'amplitude de l'onde résultante.
La diffraction, elle, est une propriété des ondes qui se caractérise par un étalement des directions de propagation de l'onde, lorsque celle-ci traverse une ouverture.

Ces deux phénomènes sont fondamentaux pour comprendre la nature ondulatoire de la lumière et d'autres types d'ondes.

## Diffraction

Le phénomène de diffraction est nettement observé lorsque la taille de l'ouverture est du même ordre de grandeur ou inférieure à la longueur d'onde.

La théorie de la diffraction, dans sa forme élémentaire, repose sur le principe de Huygens-Fresnel. Selon ce principe, chaque point atteint par une onde se comporte comme une source secondaire. La figure de diffraction observée résulte de l'interférence des ondes émises par l'ensemble des sources secondaires.

## Interférences : Les fentes d'Young

L'expérience des fentes de Young, réalisée pour la première fois par Thomas Young en 1801, est une démonstration classique du phénomène d'interférences. Cette expérience montre que la lumière se comporte comme une onde, produisant des motifs d'interférence lorsque des faisceaux lumineux sont superposés.

L'expérience consiste à faire passer une source de lumière cohérente (comme la lumière monochromatique d'un laser) à travers deux fentes très rapprochées, créant ainsi deux sources de lumière cohérentes et en phase. La lumière provenant de ces fentes se propage et se superpose sur un écran placé derrière les fentes, produisant un motif d'interférence caractéristique de franges lumineuses et sombres.


Considérons un point P sur un écran situé à une distance D  de deux fentes parallèles éclairées par une onde lumineuse de longueur d'onde λ.

La différence de chemin optique entre les rayons qui arrivent de chaque fente à ce point P est donnée par :
$$
 \delta = d \sin(\theta)
$$
Où :
- d  est la distance entre les fentes,
- θ est l'angle formé par la ligne allant des fentes au point \( P \) avec l'axe central.

La différence de phase entre les rayons provenant des deux fentes est alors :
$$
 \Delta\phi = \frac{2\pi}{\lambda} \delta = \frac{2\pi d}{\lambda} \sin(\theta)
$$
L'intensité d'interférence observée à ce point est proportionnelle au carré de la somme des amplitudes des ondes provenant de chaque fente. L'amplitude de chaque onde est proportionnelle à $ e^{i\phi} $, où $ \phi $ est la phase de l'onde.

Ainsi, l'intensité totale  I($\theta$)  est donnée par :
$$
 I(\theta) \propto \left|  e^{i \frac{\Delta\phi}{2}} + e^{-i \frac{\Delta\phi}{2}} \right|^2
$$
$$
 = \left| 2\cos\left(\frac{\Delta\phi}{2}\right) \right|^2
$$
$$
 = 4\cos^2\left(\frac{\Delta\phi}{2}\right)
$$
$$
 = 4\cos^2\left(\frac{\pi d}{\lambda} \sin(\theta) \right)
$$


Maintenant, considérons la diffraction due à la largeur finie des fentes. Cette diffraction peut être modélisée par des ouvertures rectangulaires. La transformée de Fourier de la fonction rectangulaire est la fonction sinc(x).

La distribution d'intensité due à la diffraction à travers une fente unique est proportionnelle à $ \text{sinc}^2(\beta) $, où $ \beta = \frac{\pi b}{\lambda} \sin(\theta) $, et $ b $ est la largeur de chaque fente.

Ainsi, l'intensité totale d'interférence est donnée par :
$$
 I(\theta) \propto \cos^2\left(\frac{\pi d}{\lambda} \sin(\theta) \right) \cdot \text{sinc}^2\left(\frac{\pi b}{\lambda} \sin(\theta) \right)
$$




### Simulation de l'expérience des fentes d'Young

**1. Ondes planes**

D'après le principe de Huygens-Fresnel, on peut supposer que l'expériecne d'Young est équivalente à deux sources ponctuelles d'ondes monochromatique cohérentes situées en \(0, a\) et \(0, -a\). Ainsi, la fonction d'onde en un point \(x, y\) après les fentes est une superposition des contributions provenant des deux fentes :

$$
\psi(x, y,t) = \psi_1(x, y,t) + \psi_2(x, y,t)
$$

Où

$$
\psi_1(x, y, t) = A_1 \sin \left( k \sqrt{(x - a)^2 + y^2} - \omega t  \right)$$

$$
\psi_1(x, y, t) = A_2 \sin \left( k \sqrt{(x + a)^2 + y^2} - \omega t  \right)$$


où  $A_1$  et $A_2$  sont les amplitudes des ondes passant par les fentes 1 et 2,  $k$ est le nombre d'onde (lié à la longueur d'onde $\lambda$ par $k = \frac{2\pi}{\lambda}$ ).

La fonction d'onde résultante $\psi(x,y,t)$ permet d'obtenir l'intensité finale $I(x,y)=|\psi(x,y,t)|^2$.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import interact

# Fonction pour calculer et afficher l'intensité des ondes
def plot_wave_superposition(L=20, a=1, wavelength=0.5):
    k = 2 * np.pi / wavelength  # Nombre d'onde

    # Création de la grille de coordonnées
    x = np.linspace(0, L, 1000)
    y = np.linspace(0, L, 1000)
    X, Y = np.meshgrid(x, y)

    # Positions des sources
    source1 = (0, a + L / 2)
    source2 = (0, -a + L / 2)

    # Calcul des distances des points de la grille aux sources
    R1 = np.sqrt((X - source1[0]) ** 2 + (Y - source1[1]) ** 2)
    R2 = np.sqrt((X - source2[0]) ** 2 + (Y - source2[1]) ** 2)

    # Calcul des ondes
    onde1 = np.sin(k * R1)  # onde monochromatique sinusoïdales
    onde2 = np.sin(k * R2)

    # Superposition des ondes
    onde_superposition = onde1 + onde2
    intensite = np.abs(onde_superposition) ** 2

    # Visualisation
    plt.figure(figsize=(8, 8))
    plt.imshow(intensite, extent=[0, L, 0, L], origin='lower', cmap='magma', vmax=4)
    plt.colorbar(label='Intensité')
    plt.title('Intensité - superposition de deux ondes planes')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

# Utilisation de interact pour créer les widgets de contrôle
interact(plot_wave_superposition,
         L=widgets.FloatSlider(min=5, max=30, step=1, value=20, description='L'),
         a=widgets.FloatSlider(min=0, max=5, step=0.1, value=1, description='a'),
         wavelength=widgets.FloatSlider(min=0.1, max=2, step=0.1, value=0.5, description='Longueur d\'onde')
        )
```

On observe bien les interfranges attendues dans le cadre de l'expériences des fentes d'Young.


**2. Ondes sphériques**

Les fonctions d'onde provenant des fentes peuvent être représentées comme des sources ponctuelles situées à \(0, a\) et \(0, -a\). La fonction d'onde en un point \(x, y\) après les fentes est une superposition des contributions provenant des deux fentes :

$$
\psi(x, y) = \psi_1(x, y) + \psi_2(x, y)
$$

Avec

$$
\psi_1(x, y) = A_1 \frac{e^{ik\sqrt{x^2 + (y - a)^2}}}{\sqrt{x^2 + (y - a)^2}}
$$

$$
\psi_2(x, y) = A_2 \frac{e^{ik\sqrt{x^2 + (y + a)^2}}}{\sqrt{x^2 + (y + a)^2}}
$$

où  $A_1$  et $A_2$  sont les amplitudes des ondes passant par les fentes 1 et 2.


La fonction d'onde résultante est donc :

$$
\psi(x, y) = A_1 \frac{e^{ik\sqrt{x^2 + (y - a)^2}}}{\sqrt{x^2 + (y - a)^2}} + A_2 \frac{e^{ik\sqrt{x^2 + (y + a)^2}}}{\sqrt{x^2 + (y + a)^2}}
$$

L'intensité en un point $(x, y)$ est le module carré de la fonction d'onde :

$$
I(x, y) = |\psi(x, y)|^2
$$


```python
import numpy as np
import matplotlib.pyplot as plt
import cmath

j = cmath.sqrt(-1)

# Paramètres
L = 10  # Taille du carré
a = 1.5  # Position des sources d'ondes
wavelength = 0.3  # Longueur d'onde
k = 2 * np.pi / wavelength  # Nombre d'onde

# Création de la grille de coordonnées
x = np.linspace(0, L, 500)
y = np.linspace(0, L, 500)
X, Y = np.meshgrid(x, y)

# Positions des sources
source1 = (0, a + L / 2)
source2 = (0, -a + L / 2)

# Calcul des distances des points de la grille aux sources
R1 = np.sqrt((X - source1[0]) ** 2 + (Y - source1[1]) ** 2)
R2 = np.sqrt((X - source2[0]) ** 2 + (Y - source2[1]) ** 2)

# Calcul des ondes
wave1 = np.exp(j * (k * R1)) / R1  # sin(kr)
wave2 = np.exp(j * (k * R2)) / R2

# Superposition des ondes
wave_superposition = wave1 + wave2

# Déterminer une valeur de x fixe
x_fixed = L / 2.5
index_x_fixed = np.argmin(np.abs(x - x_fixed))

# Calculer l'intensité des ondes superposées pour x fixe
intensity_fixed_x = np.abs(wave_superposition[:, index_x_fixed]) ** 2

# Création de la figure avec deux sous-figures
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 7), gridspec_kw={'width_ratios': [2, 1]})

# Visualisation de la superposition des ondes
im = ax1.imshow(np.abs(wave_superposition) ** 2, extent=[0, L, 0, L], origin='lower', cmap='magma', vmax=1)
ax1.set_title("Intensité - superposition d'ondes sphériques")
ax1.set_xlabel('x')
ax1.set_ylabel('y')
fig.colorbar(im, ax=ax1, label="Amplitude de l'onde")

# Visualisation de l'intensité pour x fixé
ax2.plot(y, intensity_fixed_x, color='blue')
ax2.set_title(f'Intensité de l\'onde pour x = {x_fixed}')
ax2.set_xlabel('y')
ax2.set_ylabel('Intensité')

plt.tight_layout()
plt.show()
```

On retrouve l'allure attendue du sinus cardinal.



# Interférences de matière

### L'hypothèse de Louis de Broglie

En 1924, Louis de Broglie fait l'hypothèse que toute matière possède une onde associée, illustrant ainsi la dualité onde-corpuscule. La relation entre la longueur d'onde et la quantité de mouvement d'un particule est donnée par la formule suivante:
\begin{equation}
\lambda = \frac{h}{p}
\end{equation}

où $h$ représente la constante de Planck.

### Les ondes de matières:  les fentes d'Young revisitées

Puisque les particules peuvent être décritent par des ondes dans certains contextes, on s'attend à observer des phénomènes d'interférences et de diffraction pour de telles ondes.


Ainsi, que se passerait-il si l'on envoyait des particules massiques à travers une fente, ou deux? Cette expérience de pensée était celle de Richard Feynmann, et fut réalisée pour la première fois en 1961 par C. Jonsson.

*Jonsson, C. (1974). Electron diffraction at multiple slits. Am. J. Phys, 42(1), 4-11.*


Encore plus remarquable, une équipe japonaise en 1989 met en évidence la construction progressive de le figure d'interférence en envoyant les électrons un par un, et non par paquets.

Ainsi, ces deux expériences mettent en évidence des figures de diffraction pour la simple fente, et les franges typiques d'interférences pour la fente double. Il semble donc que la dualité onde-corpuscule des électrons soit vérifiée expérimentalement. Cette propriété de la matière trouves diverses applications, telles que le gravimètre à interférométrie atomique.



## Explications/interprétation

<img src="https://i.ibb.co/0G0nJKv/image.png" border="0" width=500>


Soit $\Psi$ la fonction d'onde d'une particule. Lorsqu'une particule passe à travers un obstacle de type fente d'Young, sa fonction d'onde se divise et interfère avec elle-même. Cette interférence peut être constructive ou destructive, créant ainsi des zones de haute et de basse probabilité de présence de la particule. La densité de probabilité de trouver la particule en un point donné est proportionnelle à $|\Psi|^2$.
Lorsque la particule rencontre un écran de détection, elle se localise en un point précis donné par la densité de probablité. Ainsi, en envoyant de nombreuses particules, on retrouve le motif typique d'interférence.

### Limites de l'expérience et conclusion

Comme on le sait, la mesure en physique quantique peut altérer l'état d'un système. Ce principe ne fait pas exception ici: toute tentative de déterminer par quelle fente une particule passe perturbe son état quantique, et détuit le motif d'interférence: on parle "d'effondrement de la fonction d'onde. L'expérience de Young permet donc également de mettre en évidence le problème de la mesure quantique.

En conclusion, ces expériences confirment la dualité onde-corpuscule et les principes fondamentaux de la mécanique quantique, montrant que les phénomènes quantiques s'appliquent aussi bien aux particules de matière qu'aux particules de lumière.

# Simulation des ondes de matière

Dans cette partie, on souhaite observer deux phénomènes: la diffraction d'un électron à travers une fente, et les interférences résultant du passage de l'électron à travers deux fentes. Il faut pour cela disposer d'une méthode de résolution numérique de l'équation de Shcrodinger en 2D. Pour cela, plusieurs méthodes existent: nous choisirons une méthode des éléments finis, et en particulier la méthode de Cranck-Nichoson, dont la méthode et l'algorithme est tiré de https://artmenlope.github.io/solving-the-2d-schrodinger-equation-using-the-crank-nicolson-method/.

## Schéma numérique

L'équation de Schrodinger en 2D est donnée par
\begin{equation}
i \hbar \frac{\partial \psi(x, y, t)}{\partial t} = -\frac{\hbar^2}{2m} \left( \frac{\partial^2 \psi(x, y, t)}{\partial x^2} + \frac{\partial^2 \psi(x, y, t)}{\partial y^2} \right) + V(x, y, t) \psi(x, y, t)
\end{equation}

Par soucis de simplicité, on fixera $\hbar = 2m = 1$.

### Discrétisation

En utilisant la méthode de Crank-Nicolson, on discrétise les dérivées temporelles et spatiales. Soit $\psi _{i,j}^n$ la fonction d'onde à la position $(i, j)$ au au pas de temps $n$. Le schéma de Cranck-Nicholson est la moyenne d'un schéma d'Euler explicite et implicite:

\begin{equation}
i\frac{\psi_{i,j}^{n+1} - \psi_{i,j}^n}{\Delta t} = \frac{1}{2} \left( F_{i,j}^{n+1} + F_{i,j}^n \right)
\end{equation}
où
\begin{equation}
F_{i,j}^n = -\left( \frac{\psi_{i+1,j}^n - 2\psi_{i,j}^n + \psi_{i-1,j}^n}{(\Delta x)^2} + \frac{\psi_{i,j+1}^n - 2\psi_{i,j}^n + \psi_{i,j-1}^n}{(\Delta y)^2} \right) + V_{i,j} \psi_{i,j}^n
\end{equation}

Pour simplifier, nous utiliserons les constantes suivantes :
$$
r_x = \frac{\Delta t}{2i \Delta x^2}, \quad r_y = \frac{\Delta t}{2i \Delta y^2}
$$

### Formulation matricielle

Pour mettre en œuvre la méthode de Crank-Nicolson sous forme matricielle, nous devons reformuler notre équation discrétisée en termes de matrices et de vecteurs. Soit $\Psi^n$ le vecteur des valeurs de $\psi$ à l'instant $n$, et $A$ et $M$ les matrices correspondantes aux termes spatiaux du schéma Crank-Nicolson.

Le schéma Crank-Nicolson peut être écrit sous la forme suivante :

$$ A \Psi^{n+1} = M \Psi^n $$

où $A$ et $M$ sont définies comme suit.

### Construction des matrices $A$ et $M$
On pose :
$$ a_{ij} = \left(1+2r_x +2r_y + i \frac{\Delta t}{2} V_{i,j}^{n+1}\right)$$
$$ b_{ij} = \left(1-2r_x -2r_y - i \frac{\Delta t}{2} V^{n}_{i,j}\right)$$
Avec les conditions au bord on a $\psi^n_{0,j}=\psi^n_{i,0}=\psi^n_{N-1,j}=\psi^n_{i,N-1}=0$

On peut écrire le schéma de Cranck-Nicholson ainsi : $$\begin{array}{l}
    -r_y \psi^{n+1}_{i+1,j} -r_y \psi^{n+1}_{i-1,j} + a_{i,j}\psi^{n+1}_{i,j} - r_x \psi^{n+1}_{i,j+1} - r_x\psi^{n+1}_{i,j-1} = \\[2ex]
     r_y \psi^{n}_{i+1,j} +r_y \psi^{n}_{i-1,j} + b_{i,j} \psi^{n}_{i,j} + r_x \psi^{n}_{i,j+1} + r_x\psi^{n}_{i,j-1}
\end{array}$$
En utilisant cette expression on peut écrire le schéma sous forme matricielle ainsi:
$$
\textbf{A} \cdot \textbf{x} =
\begin{pmatrix}
a_{0} & -r_x & 0 & 0 &\cdots & 0 & -r_y & 0 & 0 & \cdots & 0 \\[1ex]
-r_x & a_{1} & -r_x & 0 &\cdots & & 0 & -r_y & 0 & \cdots & 0 \\[1ex]
0 & -r_x & a_{2} & -r_x &  &  &  &  & -r_y & \\[1ex]
0 & 0 & -r_x & \ddots  & \ddots  & &  &  &   &  \ddots &\\[1ex]
\vdots & \vdots &  & \ddots  &  &  & &  &  & & \\[1ex]
0&&&&&&&&&&\\[1ex]
-r_y&0&&&&&&&&&\\[1ex]
0&-r_y&&&&&&&&&\\[1ex]
0& 0&-r_y&&&&&&&\ddots&\\[1ex]
\vdots& \vdots&&\ddots&&&&&\ddots&\ddots&-r_x\\[1ex]
0&0&&&&&&&&-r_x&a_{(N-3)(N-1)}
\end{pmatrix}
\begin{pmatrix}
\psi^{n+1}_0\\[1ex]
\psi^{n+1}_1\\[1ex]
\psi^{n+1}_2\\[1ex]
\\[1ex]
\vdots \\[1ex]
\\[1ex]
\psi^{n+1}_k\\[1ex]
\\[1ex]
\vdots\\[1ex]
\\[1ex]
\psi^{n+1}_{(N-3)(N-1)}
\end{pmatrix}
$$



$$
\textbf{b} = \textbf{M} \cdot \textbf{y} =
\begin{pmatrix}
b_{0} & r_x & 0 & 0 &\cdots & 0 & r_y & 0 & 0 & \cdots & 0 \\[1ex]
r_x & b_{1} & r_x & 0 &\cdots & & 0 & r_y & 0 & \cdots & 0 \\[1ex]
0 & r_x & b_{2} & r_x &  &  &  &  & r_y & \\[1ex]
0 & 0 & r_x & \ddots  & \ddots  & &  &  &   &  \ddots &\\[1ex]
\vdots & \vdots &  & \ddots  &  &  & &  &  & & \\[1ex]
0&&&&&&&&&&\\[1ex]
r_y&0&&&&&&&&&\\[1ex]
0&r_y&&&&&&&&&\\[1ex]
0& 0&r_y&&&&&&&\ddots&\\[1ex]
\vdots& \vdots&&\ddots&&&&&\ddots&\ddots&r_x\\[1ex]
0&0&&&&&&&&r_x&b_{(N-3)(N-1)}
\end{pmatrix}
\begin{pmatrix}
\psi^{n}_0\\[1ex]
\psi^{n}_1\\[1ex]
\psi^{n}_2\\[1ex]
\\[1ex]
\vdots \\[1ex]
\\[1ex]
\psi^{n}_k\\[1ex]
\\[1ex]
\vdots\\[1ex]
\\[1ex]
\psi^{n}_{(N-3)(N-1)}
\end{pmatrix}$$





### Résolution du système

La méthode de Crank-Nicolson aboutit à un système linéaire à chaque pas de temps :

$$ A \Psi^{n+1} = M \Psi^n $$

Pour chaque itération temporelle, nous devons résoudre ce système linéaire pour obtenir $\Psi^{n+1}$.

La formulation matricielle de la méthode de Crank-Nicolson permet de transformer l'équation de Schrödinger discrétisée en un système linéaire qui peut être résolu itérativement à chaque pas de temps.

# Implémentation numérique
On souhaite mettre en pratique cette méthode, pour simuler des fonctions d'ondes dans une boîte carrée, discrétisée d'un même pas selon x et y.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from matplotlib.patches import Rectangle
from scipy.sparse import csc_matrix
from scipy.sparse.linalg import spsolve
from matplotlib.animation import FuncAnimation, FFMpegWriter

"""Système d'unité"""
hbar = 1  # Constante de Planck réduite
m = 0.5  # Masse de l'électron, 2m=1
metre = 6.67e-9 # en unités naturelles (=Longueur d'un mètre dans notre système d'unité)
g= 6.67e-24 #en unités naturelles (= valeur de g dans notre système d'unité)

""" Paramètres de la simulation"""
L = 8          # Longueur boite (carrée) en unité na
Dx = 0.08    # Pas spatial
N = int(L/Dx) + 1 # Nombre de points


# Fonction principale, simulation dans boîte carrée
def simulation(psi0,V, Dx, N, Nt=200, L=8):

    '''
       psi0: fonction d'onde initiale, taille LxL, (N-2)*(N-2) points
       V : potentiel, taille (NxN) points
       Nt: nombre d'itérations
       L: Dimension de la boîte
       Dx: Pas spatial
    '''

    Dt=Dx**2/4

    #Résolution via Cranck Nichoslon
    r = -Dt/(2j*Dx**2) # Constante utile pour la résolution
    N_ = (N-2)**2  # Nombre de facteurs inconnus pour la fonction d'onde (on rappelle que la fonction d'onde est nulle aux bords)

    #Matrice des éléments finis
    A = np.zeros((N_,N_), complex)
    M = np.zeros((N_,N_), complex)

    for k in range(N_):

        i = 1 + k//(N-2)
        j = 1 + k%(N-2)

        # Diagonale Centrale
        A[k,k] = 1 + 2*r + 2*r + 1j*Dt/2*V[i,j]
        M[k,k] = 1 - 2*r - 2*r - 1j*Dt/2*V[i,j]

        if i != 1:
            A[k,(i-2)*(N-2)+j-1] = -r
            M[k,(i-2)*(N-2)+j-1] = r

        if i != N-2:
            A[k,i*(N-2)+j-1] = -r
            M[k,i*(N-2)+j-1] = r

        if j != 1:
            A[k,k-1] = -r
            M[k,k-1] = r

        if j != N-2:
            A[k,k+1] = -r
            M[k,k+1] = r

    Asp = csc_matrix(A)

    psis = [] # Fonction d'onde à chaque itérations
    density = [] #Densité de proba à chaque itération

    psi = psi0
    psi[0,:] = psi[-1,:] = psi[:,0] = psi[:,-1] = 0 # Fonction nulle sur les points
    psis.append(np.copy(psi)) # Stockage de psi0 dans la liste
    density.append(np.copy(np.abs(psi)**2) ) # Stockage de density0 dans la liste

    #Résolution du sysème Ax=My à chaque itération
    for i in range(1,Nt):

        psi_vect = psi.reshape((N_))
        b = np.matmul(M,psi_vect) # Produit matriciel de M et psi_i
        psi_vect = spsolve(Asp,b) # Résolution du système Apsi_{i+1}=b=Mpsi_i
        psi = psi_vect.reshape((N-2,N-2))

        #Ajout aux listes
        psis.append(np.copy(psi))
        density.append(np.copy(np.abs(psi)**2))

        if i%20==0 : print(i,"/",Nt) #Pour voir la progression

    print("Simulation complete")
    return density
```

# Interférence de matière : Fentes d'Young

```python
#Paramètres de la simulation
L= 8
Dx=0.08
N=int(L/Dx)+1

"""Fonction d'onde initiale"""
k0 = 15*np.pi      # Nombre d'onde initial, arbitraire
sigma = 0.6      # Largeur du paquet d'onde initial
x0 = L/4
y0 = L/2

x = np.linspace(0, L, N-2)
y = np.linspace(0, L, N-2)
x, y = np.meshgrid(x, y)

psi0 = np.exp(-1/2*((x-x0)**2 + (y-y0)**2)/sigma**2)*np.exp(-1j*k0*(x-x0)) #Fonction d'onde initiale

density = np.float64(np.abs(psi0)**2) #Densité de proba
density /= np.sum(density) #normalisation

print("Approximation de l'énergie initiale de l'électron", hbar**2*k0**2/(2*m) ,"UA") # Sûre à 2% de cete égalité, c'est pour l'ordre de grandeur

"""Potentiel"""
v = np.zeros((N,N))

taille_trou=6
espacement=17
epaisseur=2
centre = N // 2

hole_start_1 = (N // 2) - espacement // 2 - taille_trou // 2
hole_end_1 = hole_start_1 + taille_trou
hole_start_2 = (N // 2) + espacement // 2 - taille_trou // 2
hole_end_2 = hole_start_2 + taille_trou

v[centre-epaisseur:centre+epaisseur, :] = 10**6  # Grand devant l'énergie de l'électron
v[centre-epaisseur:centre+epaisseur, hole_start_1:hole_end_1] = 0
v[centre-epaisseur:centre+epaisseur, hole_start_2:hole_end_2] = 0

v=v.T

"""Visualisation de la fonction d'onde initiale et du potentiel"""

fig, ax = plt.subplots(1, 2, figsize=(14, 6))

# Densité de probabilité
cax1 = ax[0].imshow(density, extent=(0, L, 0, L), origin='lower', cmap='magma',interpolation="gaussian")
ax[0].set_title('Densité de probabilité')
ax[0].text(0.5, -0.15, r'$1$ unit$ = 6.67 \times 10^{-9}$ m', transform=ax[0].transAxes, ha='center', fontsize=10)
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
fig.colorbar(cax1, ax=ax[0])

# Potentiel
cax2 = ax[1].imshow(v, extent=(0, L, 0, L), origin='lower', cmap='binary')
ax[1].set_title('Potentiel')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
fig.colorbar(cax2, ax=ax[1])

plt.tight_layout()
plt.show()

```
